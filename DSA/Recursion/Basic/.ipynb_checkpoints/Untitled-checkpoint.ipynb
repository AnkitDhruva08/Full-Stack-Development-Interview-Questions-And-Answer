````python
import requests
from typing import Tuple, List

# ---------- Anti-Automation Absence Check ----------

def check_anti_automation_absence(url: str, html: str) -> Tuple[int, List[str], List[str]]:
    score = 0
    issues = []
    recommendations = []

    try:
        # 1. Check for bot-blocking headers
        response = requests.get(url, timeout=10)
        headers = response.headers

        if "x-robots-tag" not in headers or "noindex" not in headers.get("x-robots-tag", "").lower():
            score += 1
        else:
            issues.append("Site blocks bots via X-Robots-Tag header.")
            recommendations.append("Remove aggressive noindex/nofollow unless essential.")

        # 2. Check for CAPTCHAs on homepage
        if "captcha" in response.text.lower():
            issues.append("Potential CAPTCHA challenge found on homepage.")
            recommendations.append("Use CAPTCHAs only on sensitive actions like signups or payments.")
        else:
            score += 2

        # 3. robots.txt check
        parsed = requests.utils.urlparse(url)
        robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
        try:
            r_txt = requests.get(robots_url, timeout=5)
            if "disallow: /" not in r_txt.text.lower():
                score += 1
            else:
                issues.append("robots.txt blocks general crawling.")
                recommendations.append("Update robots.txt to allow general crawling.")
        except:
            issues.append("robots.txt not found or unreachable.")
            recommendations.append("Ensure robots.txt is accessible and well-configured.")

        # 4. Check for fingerprinting scripts (basic heuristic)
        if any(s in response.text.lower() for s in ["fingerprintjs", "navigator.plugins", "navigator.hardwareconcurrency"]):
            issues.append("Potential fingerprinting scripts detected.")
            recommendations.append("Avoid aggressive fingerprinting that blocks automation.")
        else:
            score += 1

        return min(score, 5), issues, recommendations

    except Exception as e:
        issues.append("Anti-automation check failed.")
        recommendations.append(str(e))
        return 0, issues, recommendations
````

````python
import requests
from typing import Dict, List, Tuple
from urllib.parse import urlparse


# -----------------------------------------------------------
# ENHANCED ANTI-AUTOMATION CHECK (0‚Äì100 scoring)
# -----------------------------------------------------------

def enhanced_anti_automation_check(url: str) -> Dict:
    results = {
        "score": 0,
        "issues": [],
        "recommendations": [],
        "breakdown": {}
    }

    try:
        # ------------------------------
        # A. Multi User-Agent Checks
        # ------------------------------
        agents = {
            "chrome_desktop": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
            "mobile": "Mozilla/5.0 (Linux; Android 11)",
            "bot": "Googlebot/2.1 (+http://www.google.com/bot.html)"
        }

        status_codes = {}
        for key, agent in agents.items():
            r = requests.get(url, headers={"User-Agent": agent}, timeout=10)
            status_codes[key] = r.status_code

        if len(set(status_codes.values())) > 1:
            results["issues"].append("Different responses for different User-Agents ‚Üí bot protection likely")
            results["recommendations"].append("Standardize response for common bot UAs.")
        else:
            results["score"] += 20

        # ------------------------------
        # B. Header-Based Bot Blocking
        # ------------------------------
        r = requests.get(url, timeout=10)
        headers = r.headers
        header_score = 0

        if "x-robots-tag" in headers and "noindex" in headers["x-robots-tag"].lower():
            results["issues"].append("X-Robots-Tag contains noindex ‚Üí automation restricted.")
            results["recommendations"].append("Remove noindex unless needed.")
        else:
            header_score += 10

        results["breakdown"]["headers"] = header_score
        results["score"] += header_score

        # ------------------------------
        # C. robots.txt Check
        # ------------------------------
        parsed = urlparse(url)
        robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"

        try:
            r_txt = requests.get(robots_url, timeout=5)
            txt = r_txt.text.lower()

            if "disallow: /" in txt:
                results["issues"].append("robots.txt blocks entire website.")
                results["recommendations"].append("Remove global Disallow rule.")
            else:
                results["score"] += 10
        except:
            # Missing robots.txt is NOT an error
            results["score"] += 5
            results["breakdown"]["robots.txt"] = "missing (neutral)"

        # ------------------------------
        # D. CAPTCHA + JS challenge detection
        # ------------------------------
        body = r.text.lower()
        captcha_score = 0

        captcha_signals = ["captcha", "grecaptcha", "hcaptcha", "turnstile"]
        if any(c in body for c in captcha_signals):
            results["issues"].append("CAPTCHA detected on homepage.")
            results["recommendations"].append("Avoid CAPTCHAs on non-sensitive pages.")
        else:
            captcha_score += 20

        results["score"] += captcha_score
        results["breakdown"]["captcha"] = captcha_score

        # ------------------------------
        # E. Fingerprinting detection (stronger signals)
        # ------------------------------
        fp_signals = [
            "fingerprintjs", "canvas", "webgl", "audiocontext", "webrtc",
            "perimeterx", "human", "hcaptcha", "cloudflare"
        ]

        fp_score = 0
        if any(sig in body for sig in fp_signals):
            results["issues"].append("Advanced fingerprinting / bot manager scripts detected.")
            results["recommendations"].append("Reduce aggressive JS-based bot detection.")
        else:
            fp_score += 20

        results["score"] += fp_score
        results["breakdown"]["fingerprinting"] = fp_score

        return results

    except Exception as e:
        return {
            "score": 0,
            "issues": ["Automation check failed"],
            "recommendations": [str(e)],
            "breakdown": {}
        }
````

````python
def compare_old_new(url: str):
    print("\n=== Running OLD anti-automation check ===")
    old_score, old_issues, old_rec = check_anti_automation_absence(url, "")

    print("\n=== Running NEW enhanced anti-automation check ===")
    new_result = enhanced_anti_automation_check(url)

    comparison = {
        "old_score": old_score,
        "new_score": new_result["score"],
        "old_issues": old_issues,
        "new_issues": new_result["issues"],
        "old_recommendations": old_rec,
        "new_recommendations": new_result["recommendations"],
        "new_breakdown": new_result["breakdown"]
    }

    return comparison
````

````python
"""
Complete Working Example: Anti-Automation Script Comparison
============================================================
This script demonstrates how to run both old and new scripts,
then visualize the improvements comprehensively.
"""

import requests
from typing import Tuple, List, Dict
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Set visual style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# ==============================================================================
# OLD SCRIPT (Legacy Version)
# ==============================================================================

def check_anti_automation_absence_OLD(url: str, html: str = None) -> Tuple[int, List[str], List[str]]:
    """
    Original anti-automation check (0-5 scoring)
    """
    score = 0
    issues = []
    recommendations = []

    try:
        # 1. Check for bot-blocking headers
        response = requests.get(url, timeout=10)
        headers = response.headers

        if "x-robots-tag" not in headers or "noindex" not in headers.get("x-robots-tag", "").lower():
            score += 1
        else:
            issues.append("Site blocks bots via X-Robots-Tag header.")
            recommendations.append("Remove aggressive noindex/nofollow unless essential.")

        # 2. Check for CAPTCHAs on homepage
        if "captcha" in response.text.lower():
            issues.append("Potential CAPTCHA challenge found on homepage.")
            recommendations.append("Use CAPTCHAs only on sensitive actions like signups or payments.")
        else:
            score += 2

        # 3. robots.txt check
        parsed = requests.utils.urlparse(url)
        robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
        try:
            r_txt = requests.get(robots_url, timeout=5)
            if "disallow: /" not in r_txt.text.lower():
                score += 1
            else:
                issues.append("robots.txt blocks general crawling.")
                recommendations.append("Update robots.txt to allow general crawling.")
        except:
            issues.append("robots.txt not found or unreachable.")
            recommendations.append("Ensure robots.txt is accessible and well-configured.")

        # 4. Check for fingerprinting scripts (basic heuristic)
        if any(s in response.text.lower() for s in ["fingerprintjs", "navigator.plugins", "navigator.hardwareconcurrency"]):
            issues.append("Potential fingerprinting scripts detected.")
            recommendations.append("Avoid aggressive fingerprinting that blocks automation.")
        else:
            score += 1

        return min(score, 5), issues, recommendations

    except Exception as e:
        issues.append("Anti-automation check failed.")
        recommendations.append(str(e))
        return 0, issues, recommendations


# ==============================================================================
# NEW SCRIPT (Enhanced Version - Simplified for Demo)
# ==============================================================================

USER_AGENT_PROFILES = {
    "desktop": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120.0",
    "mobile": "Mozilla/5.0 (Linux; Android 10) AppleWebKit/537.36 Chrome/120.0 Mobile",
    "headless": "Mozilla/5.0 HeadlessChrome/120.0",
    "bot": "python-requests/2.31"
}

BOT_PROTECTION_SIGNATURES = {
    "cloudflare": ["cf-ray", "cloudflare", "__cf_bm"],
    "recaptcha": ["recaptcha", "grecaptcha", "g-recaptcha"],
    "hcaptcha": ["hcaptcha", "h-captcha"],
}


def check_anti_automation_readiness_NEW(url: str) -> Dict:
    """
    Enhanced anti-automation check (0-100 scoring with detailed analysis)
    """
    score = 100
    issues = []
    recommendations = []
    
    # Test multiple user agents
    ua_results = {}
    for ua_name, ua_string in USER_AGENT_PROFILES.items():
        try:
            response = requests.get(url, headers={"User-Agent": ua_string}, timeout=10)
            ua_results[ua_name] = {
                "status": response.status_code,
                "headers": dict(response.headers),
                "content": response.text,
                "elapsed": 0.5
            }
        except:
            ua_results[ua_name] = {"status": 0, "headers": {}, "content": "", "elapsed": 0}
    
    # Analyze UA discrimination
    bot_blocked = ua_results["bot"]["status"] in [403, 503] and ua_results["desktop"]["status"] == 200
    headless_blocked = ua_results["headless"]["status"] in [403, 503] and ua_results["desktop"]["status"] == 200
    
    if bot_blocked:
        score -= 30
        issues.append("Website blocks bot user-agents (403/503 responses)")
        recommendations.append("Allow generic bot UAs or provide developer API tokens")
    
    if headless_blocked:
        score -= 20
        issues.append("Headless browser detection active")
        recommendations.append("Relax TLS/browser fingerprint checks")
    
    # Check for bot protection services
    desktop_content = ua_results["desktop"]["content"].lower()
    desktop_headers = ua_results["desktop"]["headers"]
    
    detected_services = {}
    for service, signatures in BOT_PROTECTION_SIGNATURES.items():
        for sig in signatures:
            if sig.lower() in desktop_content or any(sig.lower() in str(v).lower() for v in desktop_headers.values()):
                detected_services[service] = {"confidence": "high", "matches": [sig]}
                score -= 15
                issues.append(f"{service.title()} bot protection detected (high confidence)")
                recommendations.append(f"Consider API access for {service.title()} bypass")
                break
    
    # Check fingerprinting
    fp_patterns = ["fingerprintjs", "canvas.todataurl", "webgl", "navigator.plugins"]
    fp_count = sum(1 for p in fp_patterns if p in desktop_content)
    
    if fp_count > 2:
        score -= 20
        issues.append(f"Advanced fingerprinting detected ({fp_count} techniques)")
        recommendations.append("Reduce fingerprinting for legitimate automation")
    
    # Rate limiting check
    if any(result["status"] == 429 for result in ua_results.values()):
        score -= 15
        issues.append("Rate limiting detected")
        recommendations.append("Increase rate limits or provide API")
    
    # Classification
    if score >= 80:
        classification = "READY"
    elif score >= 60:
        classification = "LIMITED"
    elif score >= 40:
        classification = "RESTRICTED"
    else:
        classification = "BLOCKED"
    
    return {
        "url": url,
        "score": max(score, 0),
        "classification": classification,
        "issues": issues,
        "recommendations": recommendations,
        "detailed_analysis": {
            "ua_discrimination": {
                "bot_blocked": bot_blocked,
                "headless_blocked": headless_blocked,
                "raw_results": {
                    name: {"status": data["status"], "elapsed": data["elapsed"], "redirected": False}
                    for name, data in ua_results.items()
                }
            },
            "bot_protection": detected_services,
            "fingerprinting": {
                "risk_level": "high" if fp_count > 2 else "medium" if fp_count > 0 else "low",
                "pattern_count": fp_count,
                "obfuscated_scripts": 0
            },
            "rate_limiting": {
                "detected": any(result["status"] == 429 for result in ua_results.values()),
                "indicators": []
            }
        }
    }


# ==============================================================================
# ENHANCED VISUALIZATION WITH ERROR HANDLING
# ==============================================================================

def visualize_full_analytics(old_result: dict, new_result: dict):
    """
    Enhanced visualization with backward compatibility
    """
    
    print("="*80)
    print("üß† ANTI-AUTOMATION READINESS - IMPROVEMENT ANALYTICS REPORT")
    print("="*80)
    print()
    
    # Handle different score scales
    old_score = old_result.get("score", 0)
    # If old score is 0-5, convert to 0-100 for comparison
    if old_score <= 5:
        old_score_normalized = old_score * 20
        print(f"‚ÑπÔ∏è  Note: Old score converted from 0-5 scale to 0-100 scale")
    else:
        old_score_normalized = old_score
    
    new_score = new_result.get("score", 0)
    
    # ========================================================================
    # 1. EXECUTIVE SUMMARY
    # ========================================================================
    score_improvement = new_score - old_score_normalized
    improvement_pct = (score_improvement / old_score_normalized * 100) if old_score_normalized > 0 else 0
    
    print("\nüìä EXECUTIVE SUMMARY")
    print("-" * 80)
    print(f"Old Script Score:      {old_score}/5 ‚Üí {old_score_normalized}/100 (normalized)")
    print(f"New Script Score:      {new_score}/100")
    print(f"Score Change:          {score_improvement:+.1f} points ({improvement_pct:+.1f}%)")
    print(f"Old Classification:    {old_result.get('readiness', 'N/A')}")
    print(f"New Classification:    {new_result.get('classification', 'N/A')}")
    print(f"Issues Detected (Old): {len(old_result.get('issues', []))}")
    print(f"Issues Detected (New): {len(new_result.get('issues', []))}")
    print()
    
    # ========================================================================
    # 2. SCORE VISUALIZATION
    # ========================================================================
    print("\nüìà SCORE COMPARISON")
    print("-" * 80)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Bar chart
    versions = ['Old Script\n(normalized)', 'New Script']
    scores = [old_score_normalized, new_score]
    colors = ['#ff6b6b' if s < 50 else '#ffd93d' if s < 70 else '#6bcf7f' for s in scores]
    
    bars = ax1.bar(versions, scores, color=colors, edgecolor='black', linewidth=1.5, width=0.6)
    ax1.set_ylabel('Readiness Score (0-100)', fontsize=12, fontweight='bold')
    ax1.set_title('Overall Score Comparison', fontsize=14, fontweight='bold')
    ax1.set_ylim(0, 110)
    ax1.axhline(y=80, color='green', linestyle='--', alpha=0.5, label='READY (80+)')
    ax1.axhline(y=60, color='orange', linestyle='--', alpha=0.5, label='LIMITED (60+)')
    ax1.axhline(y=40, color='red', linestyle='--', alpha=0.5, label='RESTRICTED (40+)')
    ax1.legend(loc='upper left', fontsize=9)
    ax1.grid(axis='y', alpha=0.3)
    
    # Add value labels
    for bar, score in zip(bars, scores):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height + 2,
                f'{score:.0f}', ha='center', va='bottom', fontweight='bold', fontsize=13)
    
    # Improvement gauge
    improvement_color = '#6bcf7f' if score_improvement >= 0 else '#ff6b6b'
    ax2.barh(['Score\nImprovement'], [score_improvement], color=improvement_color, 
             edgecolor='black', linewidth=1.5, height=0.3)
    ax2.set_xlabel('Point Change', fontsize=12, fontweight='bold')
    ax2.set_title('Net Improvement', fontsize=14, fontweight='bold')
    ax2.axvline(x=0, color='black', linewidth=2)
    ax2.text(score_improvement + (5 if score_improvement > 0 else -5), 0, 
             f'{score_improvement:+.0f}', va='center', fontweight='bold', fontsize=14,
             ha='left' if score_improvement > 0 else 'right')
    ax2.grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # ========================================================================
    # 3. ISSUE ANALYSIS
    # ========================================================================
    print("\n‚ö†Ô∏è  ISSUE DETECTION ANALYSIS")
    print("-" * 80)
    
    old_issues = set(old_result.get("issues", []))
    new_issues = set(new_result.get("issues", []))
    
    fixed = old_issues - new_issues
    persisting = old_issues & new_issues
    newly_detected = new_issues - old_issues
    
    print(f"\nüìä Issue Category Summary:")
    print(f"   ‚úÖ Fixed:          {len(fixed)} issue(s)")
    print(f"   ‚ö†Ô∏è  Persisting:     {len(persisting)} issue(s)")
    print(f"   üÜï Newly Detected:  {len(newly_detected)} issue(s)")
    
    # Issue breakdown chart
    fig, ax = plt.subplots(figsize=(10, 5))
    categories = ['Fixed\nIssues', 'Persisting\nIssues', 'Newly\nDetected']
    counts = [len(fixed), len(persisting), len(newly_detected)]
    colors_cat = ['#6bcf7f', '#ffd93d', '#61a5f7']
    
    bars = ax.bar(categories, counts, color=colors_cat, edgecolor='black', linewidth=1.5, width=0.5)
    ax.set_ylabel('Count', fontsize=12, fontweight='bold')
    ax.set_title('Issue Detection Breakdown', fontsize=14, fontweight='bold')
    ax.grid(axis='y', alpha=0.3)
    
    for bar, count in zip(bars, counts):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.15,
                f'{count}', ha='center', va='bottom', fontweight='bold', fontsize=13)
    
    plt.tight_layout()
    plt.show()
    
    # ========================================================================
    # 4. DETAILED TABLES
    # ========================================================================
    print("\nüìã DETAILED ISSUE COMPARISON")
    print("-" * 80)
    
    max_len = max(len(old_result.get("issues", [])), len(new_result.get("issues", [])), 1)
    old_issues_list = list(old_result.get("issues", [])) + [''] * max_len
    new_issues_list = list(new_result.get("issues", [])) + [''] * max_len
    
    issues_df = pd.DataFrame({
        "üî¥ Old Script Issues": old_issues_list[:max_len],
        "üîµ Improved Script Issues": new_issues_list[:max_len]
    })
    
    display(issues_df)
    
    # ========================================================================
    # 5. ADVANCED CAPABILITIES (if available)
    # ========================================================================
    if "detailed_analysis" in new_result:
        print("\nüöÄ ADVANCED DETECTION CAPABILITIES (New Script Only)")
        print("-" * 80)
        
        analysis = new_result["detailed_analysis"]
        
        # User-Agent Analysis
        if "ua_discrimination" in analysis and "raw_results" in analysis["ua_discrimination"]:
            ua_disc = analysis["ua_discrimination"]
            ua_data = []
            
            for agent, result in ua_disc["raw_results"].items():
                ua_data.append({
                    "User Agent": agent.title(),
                    "Status Code": result.get("status", "N/A"),
                    "Response Time": f"{result.get('elapsed', 0):.2f}s",
                    "Blocked": "üî¥ YES" if result.get("status") in [403, 503] else "üü¢ NO"
                })
            
            ua_df = pd.DataFrame(ua_data)
            
            print("\nüñ•Ô∏è  User-Agent Discrimination Test:")
            display(ua_df)
            
            print(f"\n   ‚Üí Bot UA Blocked: {'üî¥ YES' if ua_disc.get('bot_blocked') else 'üü¢ NO'}")
            print(f"   ‚Üí Headless Blocked: {'üî¥ YES' if ua_disc.get('headless_blocked') else 'üü¢ NO'}")
        
        # Bot Protection
        if "bot_protection" in analysis and analysis["bot_protection"]:
            print("\nüõ°Ô∏è  Bot Protection Services Detected:")
            for service, data in analysis["bot_protection"].items():
                conf_emoji = "üî¥" if data.get("confidence") == "high" else "üü°"
                print(f"   {conf_emoji} {service.title()}: {data.get('confidence', 'unknown')} confidence")
        
        # Fingerprinting
        if "fingerprinting" in analysis:
            fp = analysis["fingerprinting"]
            risk_emoji = "üî¥" if fp.get("risk_level") == "high" else "üü°" if fp.get("risk_level") == "medium" else "üü¢"
            print(f"\nüîç Fingerprinting Analysis:")
            print(f"   {risk_emoji} Risk Level: {fp.get('risk_level', 'unknown').upper()}")
            print(f"   ‚Üí Techniques: {fp.get('pattern_count', 0)}")
    
    print("\n" + "="*80)
    print("‚ú® ANALYSIS COMPLETE")
    print("="*80)


# ==============================================================================
# MAIN EXECUTION
# ==============================================================================

if __name__ == "__main__":
    # Test URL
    TEST_URL = "https://www.google.com"
    
    print("üîç Starting Anti-Automation Comparison Analysis")
    print(f"üìç Target URL: {TEST_URL}\n")
    
    # Run old script
    print("‚è≥ Running OLD script...")
    old_score, old_issues, old_reco = check_anti_automation_absence_OLD(TEST_URL)
    
    old_result = {
        "url": TEST_URL,
        "score": old_score,
        "readiness": "LIMITED" if old_score >= 3 else "BLOCKED",
        "issues": old_issues,
        "recommendations": old_reco
    }
    print(f"‚úÖ Old script complete: Score = {old_score}/5\n")
    
    # Run new script
    print("‚è≥ Running NEW script...")
    new_result = check_anti_automation_readiness_NEW(TEST_URL)
    print(f"‚úÖ New script complete: Score = {new_result['score']}/100\n")
    
    # Visualize comparison
    print("\n" + "="*80)
    print("üìä GENERATING COMPREHENSIVE COMPARISON REPORT")
    print("="*80 + "\n")
    
    visualize_full_analytics(old_result, new_result)
````

````python
"""
anti_automation_full.py
Single-file package containing:
 - legacy (old) anti-automation script
 - improved anti-automation readiness script
 - comparison engine to compare old vs new
 - visualization module (matplotlib + optional Plotly)

Usage:
 - Run in a Jupyter notebook or as a script.
 - Plotly visualizations require `plotly` to be installed.
 - Playwright JS checks are optional; code will skip gracefully if not available.

Example:
    python anti_automation_full.py
    # or inside a Jupyter cell:
    from anti_automation_full import run_sample_analysis
    run_sample_analysis("https://example.com", show_plotly=False)

"""

# Standard library imports
import requests
import re
import threading
from concurrent.futures import ThreadPoolExecutor, as_completed
from typing import Tuple, List, Dict, Any
from urllib.parse import urlparse
import time
import json

# Visualization imports
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
from IPython.display import display, HTML

# Optional imports
try:
    import plotly.graph_objects as go
    PLOTLY_AVAILABLE = True
except Exception:
    PLOTLY_AVAILABLE = False


# -------------------------------------------------------------------------
# ----------------------- OLD (LEGACY) SCRIPT ------------------------------
# -------------------------------------------------------------------------

def check_anti_automation_absence_OLD(url: str, html: str = None) -> Tuple[int, List[str], List[str]]:
    """Original anti-automation check (0-5 scoring). Kept unchanged for comparison.

    Returns:
        score (int 0-5), issues (List[str]), recommendations (List[str])
    """
    score = 0
    issues: List[str] = []
    recommendations: List[str] = []

    try:
        response = requests.get(url, timeout=10)
        headers = response.headers

        # 1. Check X-Robots-Tag
        if "x-robots-tag" not in headers or "noindex" not in headers.get("x-robots-tag", "").lower():
            score += 1
        else:
            issues.append("Site blocks bots via X-Robots-Tag header.")
            recommendations.append("Remove aggressive noindex/nofollow unless essential.")

        # 2. Simple CAPTCHA check
        if "captcha" in response.text.lower():
            issues.append("Potential CAPTCHA challenge found on homepage.")
            recommendations.append("Use CAPTCHAs only on sensitive actions like signups or payments.")
        else:
            score += 2

        # 3. robots.txt check
        parsed = urlparse(url)
        robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
        try:
            r_txt = requests.get(robots_url, timeout=5)
            if "disallow: /" not in r_txt.text.lower():
                score += 1
            else:
                issues.append("robots.txt blocks general crawling.")
                recommendations.append("Update robots.txt to allow general crawling.")
        except Exception:
            issues.append("robots.txt not found or unreachable.")
            recommendations.append("Ensure robots.txt is accessible and well-configured.")

        # 4. Basic fingerprinting heuristic
        if any(s in response.text.lower() for s in ["fingerprintjs", "navigator.plugins", "navigator.hardwareconcurrency"]):
            issues.append("Potential fingerprinting scripts detected.")
            recommendations.append("Avoid aggressive fingerprinting that blocks automation.")
        else:
            score += 1

        return min(score, 5), issues, recommendations

    except Exception as e:
        issues.append("Anti-automation check failed.")
        recommendations.append(str(e))
        return 0, issues, recommendations


# -------------------------------------------------------------------------
# ----------------------- IMPROVED SCRIPT ---------------------------------
# -------------------------------------------------------------------------

# User agent profiles for multi-UA testing
USER_AGENT_PROFILES = {
    "desktop": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120.0",
    "mobile": "Mozilla/5.0 (Linux; Android 10) AppleWebKit/537.36 Chrome/120.0 Mobile",
    "headless": "Mozilla/5.0 HeadlessChrome/120.0",
    "bot": "python-requests/2.31"
}

# Known bot protection signatures (expandable)
BOT_PROTECTION_SIGNATURES = {
    "cloudflare": ["cf-ray", "cloudflare", "__cf_bm", "cf-chl-bypass"],
    "recaptcha": ["recaptcha", "grecaptcha", "g-recaptcha"],
    "hcaptcha": ["hcaptcha", "h-captcha"],
    "perimeterx": ["perimeterx", "px-cf-", "px-js"],
    "datadome": ["datadome"],
    "akamai": ["akamai", "akamai-nginx"],
    "arkoselabs": ["arkose", "arkoselabs"]
}

FINGERPRINT_PATTERNS = [
    "fingerprint", "fingerprintjs", "fpjs",
    "navigator.hardwareconcurrency", "canvas.todataurl",
    "webgl", "audiocontext", "getcontext('webgl')",
    "navigator.plugins", "webrtc", "rtcpeerconnection"
]


def fetch_with_ua(url: str, ua: str, timeout: int = 10) -> Dict[str, Any]:
    """Fetch URL with specified User-Agent. Returns a dict containing status, text, headers, elapsed."""
    start = time.time()
    try:
        r = requests.get(url, headers={"User-Agent": ua}, timeout=timeout)
        elapsed = time.time() - start
        return {"status": r.status_code, "text": r.text, "headers": dict(r.headers), "elapsed": elapsed}
    except Exception:
        return {"status": 0, "text": "", "headers": {}, "elapsed": 0}


def playwright_js_check(url: str, timeout: int = 15) -> Dict[str, bool]:
    """
    Optional Playwright-based JS check. If Playwright is not installed, returns safe defaults.
    Returns:
        {"js_challenge": bool, "captcha_detected": bool}
    """
    try:
        from playwright.sync_api import sync_playwright
        result = {"js_challenge": False, "captcha_detected": False}
        with sync_playwright() as p:
            browser = p.chromium.launch(headless=True)
            page = browser.new_page(user_agent=USER_AGENT_PROFILES["desktop"])
            page.goto(url, timeout=timeout * 1000)
            html = page.content().lower()
            # heuristics
            if any(k in html for k in ["cf-challenge", "cloudflare", "verify you are human", "checking your browser"]):
                result["js_challenge"] = True
            if any(k in html for k in ["recaptcha", "hcaptcha", "turnstile", "grecaptcha"]):
                result["captcha_detected"] = True
            browser.close()
        return result
    except Exception:
        return {"js_challenge": False, "captcha_detected": False}


def analyze_robots_txt(url: str, timeout: int = 5) -> Dict[str, Any]:
    parsed = urlparse(url)
    robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
    try:
        r = requests.get(robots_url, timeout=timeout)
        text = r.text.lower()
        disallow_all = "disallow: /" in text
        has_robots = True
        # count lines with disallow
        disallow_count = text.count("disallow:")
        return {"exists": has_robots, "disallow_all": disallow_all, "disallow_count": disallow_count, "raw": text}
    except Exception:
        return {"exists": False, "disallow_all": False, "disallow_count": 0, "raw": ""}


def detect_bot_protection(desktop_html: str, desktop_headers: Dict[str, str]) -> Dict[str, Dict[str, Any]]:
    detected = {}
    lower_html = desktop_html.lower()
    for name, sigs in BOT_PROTECTION_SIGNATURES.items():
        for s in sigs:
            if s in lower_html or any(s in str(v).lower() for v in desktop_headers.values()):
                detected[name] = {"confidence": "high", "matches": [s]}
                break
    return detected


def analyze_fingerprinting(desktop_html: str) -> Dict[str, Any]:
    lower = desktop_html.lower()
    count = sum(1 for p in FINGERPRINT_PATTERNS if p in lower)
    risk = "low"
    if count >= 4:
        risk = "high"
    elif count >= 1:
        risk = "medium"
    return {"pattern_count": count, "risk_level": risk}


def detect_rate_limiting(ua_results: Dict[str, Dict[str, Any]]) -> Dict[str, Any]:
    found = any(res.get("status") == 429 for res in ua_results.values())
    headers_indicators = []
    for name, res in ua_results.items():
        hdrs = res.get("headers", {})
        for k in hdrs:
            if k.lower().startswith("x-ratelimit") or k.lower() in ("retry-after",):
                headers_indicators.append({"ua": name, "header": k, "value": hdrs.get(k)})
    return {"detected": found, "indicators": headers_indicators}


def check_anti_automation_readiness_NEW(url: str, use_playwright: bool = False, parallel: bool = True) -> Dict[str, Any]:
    """Main improved function returning a structured dictionary (0-100 score).

    Parameters:
        url: target website
        use_playwright: if True, attempt Playwright JS checks
        parallel: if True, run UA checks in parallel
    """
    # Start with perfect score
    score = 100
    issues: List[str] = []
    recommendations: List[str] = []

    # 1) Multi-UA testing
    ua_results: Dict[str, Dict[str, Any]] = {}

    if parallel:
        with ThreadPoolExecutor(max_workers=len(USER_AGENT_PROFILES)) as ex:
            futures = {ex.submit(fetch_with_ua, url, ua): name for name, ua in USER_AGENT_PROFILES.items()}
            for fut in as_completed(futures):
                name = futures[fut]
                ua_results[name] = fut.result()
    else:
        for name, ua in USER_AGENT_PROFILES.items():
            ua_results[name] = fetch_with_ua(url, ua)

    # 2) UA discrimination checks
    try:
        bot_status = ua_results.get("bot", {}).get("status", 0)
        desktop_status = ua_results.get("desktop", {}).get("status", 0)
        headless_status = ua_results.get("headless", {}).get("status", 0)
    except Exception:
        bot_status = desktop_status = headless_status = 0

    # Bot UA blocked
    if bot_status in (403, 503) and desktop_status == 200:
        score -= 30
        issues.append("Website blocks automated/bot user-agents.")
        recommendations.append("Allow generic bot user-agent or provide developer token/API.")

    # Headless browser blocked
    if headless_status in (403, 503) and desktop_status == 200:
        score -= 20
        issues.append("Headless browser traffic blocked (possible fingerprint detection).")
        recommendations.append("Relax headless detection or whitelist trusted agents.")

    # 3) Rate limit detection
    rate_info = detect_rate_limiting(ua_results)
    if rate_info.get("detected"):
        score -= 15
        issues.append("Website rate-limits automated traffic (HTTP 429 detected).")
        recommendations.append("Increase rate-limit threshold or provide authenticated API access.")

    # 4) Bot protection & fingerprinting
    desktop_html = ua_results.get("desktop", {}).get("text", "")
    desktop_headers = ua_results.get("desktop", {}).get("headers", {})

    bot_protection = detect_bot_protection(desktop_html, desktop_headers)
    if bot_protection:
        # penalize based on number of services and confidence
        penalty = min(25, 15 * len(bot_protection))
        score -= penalty
        for k, v in bot_protection.items():
            issues.append(f"{k.title()} bot protection detected ({v.get('confidence', 'unknown')} confidence).")
            recommendations.append(f"Consider API access or service-specific integration for {k.title()}.")

    fp = analyze_fingerprinting(desktop_html)
    if fp.get("pattern_count", 0) >= 4:
        score -= 20
        issues.append(f"Advanced fingerprinting detected ({fp['pattern_count']} techniques).")
        recommendations.append("Reduce fingerprinting scripts for legitimate automation flows.")
    elif fp.get("pattern_count", 0) > 0:
        score -= 5
        issues.append(f"Fingerprinting indicators found ({fp['pattern_count']} patterns).")
        recommendations.append("Monitor and reduce fingerprinting where possible.")

    # 5) robots.txt analysis
    robots = analyze_robots_txt(url)
    if robots.get("exists"):
        if robots.get("disallow_all"):
            score -= 10
            issues.append("robots.txt heavily restricts crawling (Disallow: /).")
            recommendations.append("Relax robots.txt for general discovery or provide sitemap/API.")
    else:
        # Missing robots.txt is neutral but recommend adding one
        issues.append("robots.txt missing or unreachable.")
        recommendations.append("Provide a minimal robots.txt for predictable agent behavior.")

    # 6) Basic (static) CAPTCHA detection
    if "captcha" in desktop_html.lower():
        score -= 10
        issues.append("Visible CAPTCHA detected on content pages.")
        recommendations.append("Use CAPTCHAs sparingly; avoid on general content pages.")

    # 7) Optional Playwright JS checks
    js_checks = {"js_challenge": False, "captcha_detected": False}
    if use_playwright:
        js_checks = playwright_js_check(url)
        if js_checks.get("js_challenge"):
            score -= 15
            issues.append("JavaScript-based bot challenge detected.")
            recommendations.append("Reduce challenge sensitivity or whitelist trusted agents.")
        if js_checks.get("captcha_detected"):
            score -= 10
            issues.append("JS-rendered CAPTCHA detected.")
            recommendations.append("Avoid invisible/JS CAPTCHAs on general pages.")

    # 8) Header checks (X-Robots-Tag)
    desktop_headers_lower = {k.lower(): str(v).lower() for k, v in desktop_headers.items()}
    if "x-robots-tag" in desktop_headers_lower and any(tok in desktop_headers_lower.get("x-robots-tag", "") for tok in ("noindex", "nofollow")):
        score -= 5
        issues.append("X-Robots-Tag header blocks indexing for certain agents.")
        recommendations.append("Only set noindex/nofollow on sensitive pages.")

    # Ensure score bounds
    score = max(min(score, 100), 0)

    # Readiness classification
    if score >= 80:
        readiness = "READY"
    elif score >= 60:
        readiness = "LIMITED"
    elif score >= 40:
        readiness = "RESTRICTED"
    else:
        readiness = "BLOCKED"

    return {
        "url": url,
        "score": score,
        "readiness": readiness,
        "issues": issues,
        "recommendations": recommendations,
        "ua_analysis": ua_results,
        "bot_protection": bot_protection,
        "fingerprinting": fp,
        "robots": robots,
        "rate_limit": rate_info,
        "js_checks": js_checks
    }


# -------------------------------------------------------------------------
# ----------------------- COMPARISON ENGINE --------------------------------
# -------------------------------------------------------------------------


def compare_old_new(old_res: Dict[str, Any], new_res: Dict[str, Any]) -> Dict[str, Any]:
    """Compare outputs and produce a delta report."""
    # Normalize old score to 0-100 if needed
    old_score = old_res.get("score", 0)
    if old_score <= 5:
        old_score_normalized = old_score * 20
    else:
        old_score_normalized = old_score

    new_score = new_res.get("score", 0)
    score_change = new_score - old_score_normalized

    old_issues = set(old_res.get("issues", []))
    new_issues = set(new_res.get("issues", []))

    fixed = list(old_issues - new_issues)
    persisting = list(old_issues & new_issues)
    newly_detected = list(new_issues - old_issues)

    # UA comparison summary
    ua_old = {"desktop": None}
    ua_new = {}
    try:
        for k, v in new_res.get("ua_analysis", {}).items():
            ua_new[k] = v.get("status")
    except Exception:
        ua_new = {}

    summary = {
        "old_score_normalized": old_score_normalized,
        "new_score": new_score,
        "score_change": score_change,
        "fixed_issues": fixed,
        "persisting_issues": persisting,
        "newly_detected_issues": newly_detected,
        "ua_status_new": ua_new
    }
    return summary


# -------------------------------------------------------------------------
# ----------------------- VISUALIZATION MODULE -----------------------------
# -------------------------------------------------------------------------


def visualize_matplotlib_comparison(old_res: Dict[str, Any], new_res: Dict[str, Any]):
    """Matplotlib + pandas based visualization. Suitable for notebooks."""
    # Normalize old score
    old_score = old_res.get("score", 0)
    if old_score <= 5:
        old_score_norm = old_score * 20
    else:
        old_score_norm = old_score

    new_score = new_res.get("score", 0)

    # A: Score comparison bar
    plt.figure(figsize=(8, 4))
    versions = ['Old (norm)', 'New']
    scores = [old_score_norm, new_score]
    colors = ['#ff6b6b' if s < 50 else '#ffd93d' if s < 70 else '#6bcf7f' for s in scores]
    bars = plt.bar(versions, scores, color=colors, edgecolor='black')
    plt.ylim(0, 110)
    plt.title('Readiness Score: Old vs New')
    plt.ylabel('Score (0-100)')
    for bar, s in zip(bars, scores):
        plt.text(bar.get_x() + bar.get_width()/2, s + 2, f'{s:.0f}', ha='center', fontweight='bold')
    plt.axhline(80, color='green', linestyle='--', alpha=0.6)
    plt.axhline(60, color='orange', linestyle='--', alpha=0.4)
    plt.show()

    # B: Issue comparison table
    old_issues = list(old_res.get('issues', []))
    new_issues = list(new_res.get('issues', []))
    max_len = max(len(old_issues), len(new_issues), 1)
    old_issues += [''] * (max_len - len(old_issues))
    new_issues += [''] * (max_len - len(new_issues))
    issues_df = pd.DataFrame({'Old Script Issues': old_issues, 'Improved Script Issues': new_issues})
    print('\nIssue Comparison (Old vs Improved)')
    display(issues_df)

    # C: Fix matrix
    old_set = set(old_res.get('issues', []))
    new_set = set(new_res.get('issues', []))
    fixed = old_set - new_set
    persisting = old_set & new_set
    newly = new_set - old_set
    fix_df = pd.DataFrame({'Category': ['Fixed', 'Persisting', 'Newly Detected'], 'Count': [len(fixed), len(persisting), len(newly)],
                            'Items': [list(fixed) if fixed else ['None'], list(persisting) if persisting else ['None'], list(newly) if newly else ['None']]})
    print('\nIssue Fix Matrix')
    display(fix_df)

    # D: UA behaviour table
    ua = new_res.get('ua_analysis', {})
    ua_rows = []
    for name, data in ua.items():
        ua_rows.append({'User Agent': name, 'Status': data.get('status'), 'Elapsed (s)': round(data.get('elapsed', 0), 2)})
    if ua_rows:
        ua_df = pd.DataFrame(ua_rows)
        print('\nUser-Agent Analysis (New Script)')
        display(ua_df)

    # E: Radar chart for capability coverage (old vs new)
    categories = ['Bot Headers', 'CAPTCHA Detection', 'Robots.txt', 'Fingerprinting', 'JS Challenge', 'Rate-Limit', 'Multi-UA']
    N = len(categories)
    # old capability vector (best-effort)
    old_vals = [1, 1, 1, 1, 0, 0, 0]
    new_vals = [1, 1 if 'captcha' in ' '.join(new_res.get('issues', [])).lower() else 1, 1 if new_res.get('robots', {}).get('exists', False) else 1,
                1 if new_res.get('fingerprinting', {}).get('pattern_count', 0) > 0 else 0, 1 if new_res.get('js_checks', {}).get('js_challenge') else 0,
                1 if new_res.get('rate_limit', {}).get('detected') else 0, 1]

    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    old_vals_cycle = old_vals + old_vals[:1]
    new_vals_cycle = new_vals + new_vals[:1]

    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, polar=True)
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
    plt.xticks(angles[:-1], categories)
    ax.plot(angles, old_vals_cycle, linewidth=1, linestyle='--', label='Old')
    ax.fill(angles, old_vals_cycle, alpha=0.1)
    ax.plot(angles, new_vals_cycle, linewidth=2, label='New')
    ax.fill(angles, new_vals_cycle, alpha=0.2)
    plt.title('Capability Coverage (Old vs New)')
    plt.legend(loc='upper right')
    plt.show()


def visualize_plotly_comparison(old_res: Dict[str, Any], new_res: Dict[str, Any]):
    """Interactive Plotly dashboard. Requires plotly library."""
    if not PLOTLY_AVAILABLE:
        print("Plotly not installed. Install `plotly` to use this visualization.")
        return

    # Normalize
    old_score = old_res.get('score', 0)
    if old_score <= 5:
        old_score_norm = old_score * 20
    else:
        old_score_norm = old_score
    new_score = new_res.get('score', 0)

    # Score bar
    fig = go.Figure()
    fig.add_trace(go.Bar(x=['Old (norm)'], y=[old_score_norm], name='Old'))
    fig.add_trace(go.Bar(x=['New'], y=[new_score], name='New'))
    fig.update_layout(title='Readiness Score Comparison', yaxis=dict(range=[0, 110]))

    # Issues table
    old_issues = old_res.get('issues', [])
    new_issues = new_res.get('issues', [])
    max_len = max(len(old_issues), len(new_issues), 1)
    old_issues += [''] * (max_len - len(old_issues))
    new_issues += [''] * (max_len - len(new_issues))

    table = go.Figure(data=[go.Table(header=dict(values=['Old Script Issues', 'Improved Script Issues']),
                                     cells=dict(values=[old_issues, new_issues]))])

    # UA table
    ua = new_res.get('ua_analysis', {})
    ua_names = []
    ua_status = []
    ua_elapsed = []
    for k, v in ua.items():
        ua_names.append(k)
        ua_status.append(v.get('status'))
        ua_elapsed.append(round(v.get('elapsed', 0), 2))

    ua_table = go.Figure(data=[go.Table(header=dict(values=['User Agent', 'Status', 'Elapsed (s)']),
                                        cells=dict(values=[ua_names, ua_status, ua_elapsed]))])

    # Show figures
    fig.show()
    table.show()
    ua_table.show()


# -------------------------------------------------------------------------
# ----------------------- SAMPLE / MAIN RUNNER -----------------------------
# -------------------------------------------------------------------------


def run_sample_analysis(target_url: str = "https://www.google.com", use_playwright: bool = False, show_plotly: bool = False):
    """Helper to run the old and new scripts against a target and visualize results.

    Parameters:
        target_url: URL to test
        use_playwright: whether to attempt Playwright JS checks
        show_plotly: whether to show Plotly interactive dashboard (requires plotly)
    """
    print(f"Running legacy (old) script against: {target_url}")
    old_score, old_issues, old_reco = check_anti_automation_absence_OLD(target_url)
    old_result = {
        "url": target_url,
        "score": old_score,
        "readiness": "LIMITED" if old_score >= 3 else "BLOCKED",
        "issues": old_issues,
        "recommendations": old_reco
    }
    print(f"Old script: score={old_score}/5, issues={len(old_issues)}")

    print(f"\nRunning improved (new) script against: {target_url}")
    new_result = check_anti_automation_readiness_NEW(target_url, use_playwright=use_playwright, parallel=True)
    print(f"New script: score={new_result['score']}/100, readiness={new_result['readiness']}, issues={len(new_result['issues'])}")

    print("\nComparing results...")
    summary = compare_old_new(old_result, new_result)

    print('\nGenerating Matplotlib visual report...')
    visualize_matplotlib_comparison(old_result, new_result)

    if show_plotly:
        print('\nGenerating Plotly interactive report...')
        visualize_plotly_comparison(old_result, new_result)

    # Also return the raw results for programmatic use
    return {"old": old_result, "new": new_result, "summary": summary}


if __name__ == "__main__":
    # Example run (adjust TEST_URL to any site you want to test)
    TEST_URL = "https://www.google.com"
    print("Anti-Automation Full Single-File Script")
    print("====================================")
    results = run_sample_analysis(TEST_URL, use_playwright=False, show_plotly=False)
    print('\nResults summary:')
    print(json.dumps(results['summary'], indent=2))
````

````python
"""
üöÄ ENHANCED ANTI-AUTOMATION READINESS MODULE
Complete Production-Ready Implementation with Advanced Detection Capabilities
"""

import requests
import time
import re
import json
from typing import Dict, List, Tuple, Optional, Any
from urllib.parse import urlparse
from concurrent.futures import ThreadPoolExecutor, as_completed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from enum import Enum

# ==============================================================================
# ENHANCED CONFIGURATION & CONSTANTS
# ==============================================================================

class RiskLevel(Enum):
    LOW = "low"
    MEDIUM = "medium" 
    HIGH = "high"
    CRITICAL = "critical"

@dataclass
class ProtectionService:
    name: str
    signatures: List[str]
    headers: List[str]
    risk_level: RiskLevel
    bypass_recommendation: str

# Enhanced User Agent Profiles
USER_AGENT_PROFILES = {
    "desktop_chrome": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "desktop_firefox": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:121.0) Gecko/20100101 Firefox/121.0",
    "mobile_ios": "Mozilla/5.0 (iPhone; CPU iPhone OS 17_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.2 Mobile/15E148 Safari/604.1",
    "headless_chrome": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) HeadlessChrome/120.0.0.0 Safari/537.36",
    "python_bot": "python-requests/2.31.0",
    "legacy_browser": "Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko"
}

# Comprehensive Bot Protection Services Database
BOT_PROTECTION_SERVICES = {
    "cloudflare": ProtectionService(
        name="Cloudflare Bot Management",
        signatures=["cloudflare", "cf-ray", "__cf_bm", "cf-chl-bypass", "challenge-platform"],
        headers=["cf-ray", "server", "cf-cache-status"],
        risk_level=RiskLevel.HIGH,
        bypass_recommendation="Use Cloudflare API tokens or residential proxies"
    ),
    "perimeterx": ProtectionService(
        name="PerimeterX/HUMAN Security",
        signatures=["perimeterx", "pxinit", "pxcaptcha", "humansecurity"],
        headers=["x-px", "server"],
        risk_level=RiskLevel.HIGH,
        bypass_recommendation="Requires behavioral analysis bypass or mobile emulation"
    ),
    "datadome": ProtectionService(
        name="DataDome",
        signatures=["datadome", "ddg.js", "ddg_bot", "ddg_mobile"],
        headers=["x-datadome", "server"],
        risk_level=RiskLevel.HIGH,
        bypass_recommendation="Use Datadome API or advanced fingerprint rotation"
    ),
    "akamai": ProtectionService(
        name="Akamai Bot Manager",
        signatures=["akamai", "ak_bm", "_abck", "akacd"],
        headers=["x-akamai-transformed", "server"],
        risk_level=RiskLevel.MEDIUM,
        bypass_recommendation="Akamai API access or session replay techniques"
    ),
    "recaptcha": ProtectionService(
        name="Google reCAPTCHA",
        signatures=["recaptcha", "grecaptcha", "g-recaptcha", "recaptcha/api"],
        headers=[],
        risk_level=RiskLevel.MEDIUM,
        bypass_recommendation="reCAPTCHA solver services or enterprise API"
    ),
    "hcaptcha": ProtectionService(
        name="hCaptcha",
        signatures=["hcaptcha", "h-captcha", "hcaptcha.com"],
        headers=[],
        risk_level=RiskLevel.MEDIUM,
        bypass_recommendation="hCaptcha solver services or enterprise bypass"
    ),
    "arkose": ProtectionService(
        name="Arkose Labs/FunCAPTCHA",
        signatures=["arkoselabs", "funcaptcha", "arkose"],
        headers=[],
        risk_level=RiskLevel.HIGH,
        bypass_recommendation="Arkose API or advanced captcha solving"
    ),
    "shape": ProtectionService(
        name="Shape Security",
        signatures=["shapesecurity", "shape_rc", "ss_rc"],
        headers=["x-shape-request-id"],
        risk_level=RiskLevel.HIGH,
        bypass_recommendation="Requires mobile device emulation"
    )
}

# Advanced Fingerprinting Patterns
FINGERPRINTING_PATTERNS = {
    "canvas": [
        r"canvas\.(getContext|toDataURL|toBlob)",
        r"fillText.*fingerprint",
        r"getImageData"
    ],
    "webgl": [
        r"webgl",
        r"getParameter.*webgl",
        r"renderer.*webgl",
        r"getSupportedExtensions"
    ],
    "audio": [
        r"audiocontext",
        r"createanalyser",
        r"getchanneldata"
    ],
    "hardware": [
        r"hardwareconcurrency",
        r"devicememory",
        r"maxTouchPoints"
    ],
    "screen": [
        r"screen\.(width|height|avail)",
        r"pixelDepth",
        r"colorDepth"
    ],
    "plugins": [
        r"navigator\.plugins",
        r"navigator\.mimeTypes"
    ],
    "webrtc": [
        r"rtcpeerconnection",
        r"createDataChannel",
        r"getStats"
    ],
    "fonts": [
        r"getComputedStyle",
        r"fonts\.check",
        r"measureText"
    ],
    "battery": [
        r"getBattery",
        r"battery.*level"
    ],
    "obfuscated": [
        r"eval.*function",
        r"escape.*unescape",
        r"fromCharCode",
        r"decodeURIComponent"
    ]
}

# Rate Limiting Headers
RATE_LIMIT_HEADERS = {
    "x-ratelimit-limit": "Request limit",
    "x-ratelimit-remaining": "Remaining requests", 
    "x-ratelimit-reset": "Reset time",
    "retry-after": "Retry delay",
    "x-rate-limit-limit": "Alternative limit header"
}

# ==============================================================================
# CORE DETECTION ENGINE
# ==============================================================================

class AntiAutomationDetector:
    """
    Advanced anti-automation detection engine with multi-layered analysis
    """
    
    def __init__(self, timeout: int = 15, max_workers: int = 4):
        self.timeout = timeout
        self.max_workers = max_workers
        self.session = requests.Session()
        
    def fetch_with_ua(self, url: str, user_agent: str) -> Dict[str, Any]:
        """Fetch URL with specific user agent and comprehensive metrics"""
        try:
            start_time = time.time()
            response = self.session.get(
                url, 
                headers={"User-Agent": user_agent},
                timeout=self.timeout,
                allow_redirects=True
            )
            elapsed = time.time() - start_time
            
            return {
                "status_code": response.status_code,
                "headers": dict(response.headers),
                "content": response.text,
                "elapsed_time": elapsed,
                "redirect_count": len(response.history),
                "final_url": response.url,
                "content_length": len(response.text),
                "success": True
            }
        except Exception as e:
            return {
                "status_code": 0,
                "headers": {},
                "content": "",
                "elapsed_time": 0,
                "redirect_count": 0,
                "final_url": url,
                "content_length": 0,
                "success": False,
                "error": str(e)
            }
    
    def test_user_agent_discrimination(self, url: str) -> Dict[str, Any]:
        """Test multiple user agents in parallel for discrimination patterns"""
        print(f"üîç Testing {len(USER_AGENT_PROFILES)} user agents in parallel...")
        
        with ThreadPoolExecutor(max_workers=self.max_workers) as executor:
            future_to_ua = {
                executor.submit(self.fetch_with_ua, url, ua_string): ua_name
                for ua_name, ua_string in USER_AGENT_PROFILES.items()
            }
            
            results = {}
            for future in as_completed(future_to_ua):
                ua_name = future_to_ua[future]
                results[ua_name] = future.result()
        
        # Analyze discrimination patterns
        desktop_ok = results["desktop_chrome"]["status_code"] == 200
        bot_blocked = results["python_bot"]["status_code"] in [403, 503, 429]
        headless_blocked = results["headless_chrome"]["status_code"] in [403, 503, 429]
        mobile_diff = results["mobile_ios"]["status_code"] != results["desktop_chrome"]["status_code"]
        legacy_diff = results["legacy_browser"]["status_code"] != results["desktop_chrome"]["status_code"]
        
        return {
            "raw_results": results,
            "discrimination_analysis": {
                "bot_blocked": bot_blocked and desktop_ok,
                "headless_blocked": headless_blocked and desktop_ok,
                "mobile_different": mobile_diff,
                "legacy_different": legacy_diff,
                "response_time_variance": self._calculate_response_time_variance(results)
            }
        }
    
    def detect_bot_protection_services(self, content: str, headers: Dict) -> Dict[str, Any]:
        """Comprehensive bot protection service detection"""
        detected_services = {}
        content_lower = content.lower()
        
        for service_id, service in BOT_PROTECTION_SERVICES.items():
            # Check content signatures
            content_matches = []
            for signature in service.signatures:
                if signature.lower() in content_lower:
                    content_matches.append(signature)
            
            # Check header signatures
            header_matches = []
            for header_key, header_value in headers.items():
                header_str = f"{header_key}: {header_value}".lower()
                for signature in service.signatures + service.headers:
                    if signature.lower() in header_str:
                        header_matches.append(f"{header_key}: {signature}")
            
            if content_matches or header_matches:
                confidence = "high" if len(content_matches) > 2 or len(header_matches) > 1 else "medium"
                
                detected_services[service_id] = {
                    "service_name": service.name,
                    "confidence": confidence,
                    "risk_level": service.risk_level.value,
                    "content_matches": content_matches,
                    "header_matches": header_matches,
                    "evidence_count": len(content_matches) + len(header_matches),
                    "bypass_recommendation": service.bypass_recommendation
                }
        
        return detected_services
    
    def analyze_fingerprinting(self, content: str) -> Dict[str, Any]:
        """Advanced fingerprinting technique detection"""
        fingerprint_analysis = {}
        content_lower = content.lower()
        
        for technique, patterns in FINGERPRINTING_PATTERNS.items():
            detected_patterns = []
            for pattern in patterns:
                matches = re.findall(pattern, content_lower, re.IGNORECASE)
                detected_patterns.extend(matches)
            
            if detected_patterns:
                fingerprint_analysis[technique] = {
                    "pattern_count": len(detected_patterns),
                    "detected_patterns": detected_patterns[:5],  # Limit output
                    "risk_level": self._assess_fingerprinting_risk(technique, len(detected_patterns))
                }
        
        # Calculate overall fingerprinting risk
        total_patterns = sum(data["pattern_count"] for data in fingerprint_analysis.values())
        obfuscated_count = len(fingerprint_analysis.get("obfuscated", {}).get("detected_patterns", []))
        
        overall_risk = "low"
        if total_patterns > 10 or obfuscated_count > 3:
            overall_risk = "high"
        elif total_patterns > 5:
            overall_risk = "medium"
        
        return {
            "techniques_detected": fingerprint_analysis,
            "summary": {
                "total_techniques": len(fingerprint_analysis),
                "total_patterns": total_patterns,
                "obfuscated_patterns": obfuscated_count,
                "overall_risk": overall_risk
            }
        }
    
    def analyze_robots_txt(self, url: str) -> Dict[str, Any]:
        """Intelligent robots.txt analysis with context awareness"""
        parsed = urlparse(url)
        robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
        
        try:
            response = self.session.get(robots_url, timeout=10)
            if response.status_code != 200:
                return {
                    "exists": False,
                    "accessible": False,
                    "restrictions": [],
                    "risk_level": "low",
                    "analysis": "No robots.txt found or not accessible"
                }
            
            content = response.text
            restrictions = self._parse_robots_txt(content)
            
            # Risk assessment
            risk_level = "low"
            if any("disallow: /" in line.lower() for line in restrictions):
                risk_level = "high"
            elif len(restrictions) > 5:
                risk_level = "medium"
            
            return {
                "exists": True,
                "accessible": True,
                "content": content,
                "restrictions": restrictions,
                "restriction_count": len(restrictions),
                "risk_level": risk_level,
                "analysis": self._generate_robots_analysis(restrictions)
            }
            
        except Exception as e:
            return {
                "exists": False,
                "accessible": False,
                "error": str(e),
                "risk_level": "low",
                "analysis": "robots.txt check failed"
            }
    
    def detect_rate_limiting(self, headers: Dict, status_code: int) -> Dict[str, Any]:
        """Comprehensive rate limiting detection"""
        rate_limit_indicators = {}
        
        # Header-based detection
        for header, description in RATE_LIMIT_HEADERS.items():
            if header in headers:
                rate_limit_indicators[header] = {
                    "value": headers[header],
                    "description": description
                }
        
        # Status code detection
        rate_detected = status_code == 429
        if status_code == 429:
            rate_limit_indicators["status_code"] = {
                "value": "429 Too Many Requests",
                "description": "Explicit rate limiting response"
            }
        
        # Assess severity
        severity = "none"
        if rate_detected:
            severity = "high"
        elif len(rate_limit_indicators) > 2:
            severity = "medium"
        elif len(rate_limit_indicators) > 0:
            severity = "low"
        
        return {
            "detected": len(rate_limit_indicators) > 0,
            "severity": severity,
            "indicators": rate_limit_indicators,
            "recommendation": self._generate_rate_limit_recommendation(severity)
        }
    
    def analyze_javascript_challenges(self, content: str, headers: Dict) -> Dict[str, Any]:
        """Detect JavaScript-based challenges and protections"""
        js_analysis = {}
        content_lower = content.lower()
        
        # Cloudflare challenge detection
        cloudflare_indicators = [
            "checking your browser",
            "ddos protection",
            "please wait",
            "cloudflare ray id",
            "jschl-answer"
        ]
        
        cf_matches = [ind for ind in cloudflare_indicators if ind in content_lower]
        if cf_matches:
            js_analysis["cloudflare_challenge"] = {
                "detected": True,
                "indicators": cf_matches,
                "risk_level": "high"
            }
        
        # JS-based redirects and challenges
        js_redirect_patterns = [
            r"window\.location\.replace",
            r"window\.location\.href.*=.*challenge",
            r"setTimeout.*redirect",
            r"document\.cookie.*challenge"
        ]
        
        redirect_matches = []
        for pattern in js_redirect_patterns:
            matches = re.findall(pattern, content_lower)
            redirect_matches.extend(matches)
        
        if redirect_matches:
            js_analysis["javascript_redirects"] = {
                "detected": True,
                "patterns": redirect_matches,
                "risk_level": "medium"
            }
        
        # Obfuscated JS detection
        suspicious_js_patterns = [
            r"eval\(.*atob",
            r"document\.write.*fromCharCode",
            r"setInterval.*decode",
            r"function.*constructor"
        ]
        
        obfuscated_matches = []
        for pattern in suspicious_js_patterns:
            matches = re.findall(pattern, content_lower)
            obfuscated_matches.extend(matches)
        
        if obfuscated_matches:
            js_analysis["obfuscated_javascript"] = {
                "detected": True,
                "pattern_count": len(obfuscated_matches),
                "risk_level": "medium"
            }
        
        return js_analysis
    
    # ==========================================================================
    # HELPER METHODS
    # ==========================================================================
    
    def _calculate_response_time_variance(self, results: Dict) -> float:
        """Calculate variance in response times across user agents"""
        times = [data["elapsed_time"] for data in results.values() if data["success"]]
        if len(times) < 2:
            return 0.0
        mean = sum(times) / len(times)
        variance = sum((t - mean) ** 2 for t in times) / len(times)
        return variance
    
    def _assess_fingerprinting_risk(self, technique: str, pattern_count: int) -> str:
        """Assess risk level for fingerprinting techniques"""
        high_risk_tech = ["canvas", "webgl", "audio", "obfuscated"]
        medium_risk_tech = ["hardware", "webrtc", "fonts"]
        
        if technique in high_risk_tech and pattern_count > 1:
            return "high"
        elif technique in medium_risk_tech and pattern_count > 2:
            return "medium"
        elif pattern_count > 3:
            return "medium"
        else:
            return "low"
    
    def _parse_robots_txt(self, content: str) -> List[str]:
        """Parse robots.txt for meaningful restrictions"""
        restrictions = []
        lines = content.split('\n')
        
        for line in lines:
            line = line.strip()
            if line and not line.startswith('#'):
                restrictions.append(line)
        
        return restrictions
    
    def _generate_robots_analysis(self, restrictions: List[str]) -> str:
        """Generate intelligent analysis of robots.txt restrictions"""
        if not restrictions:
            return "No significant restrictions found"
        
        disallow_count = sum(1 for r in restrictions if 'disallow:' in r.lower())
        allow_count = sum(1 for r in restrictions if 'allow:' in r.lower())
        
        if disallow_count == 0:
            return "Permissive robots.txt - no disallow rules"
        elif any('disallow: /' in r.lower() for r in restrictions):
            return "Restrictive robots.txt - blocks entire site"
        elif disallow_count > 10:
            return "Comprehensive robots.txt - many specific restrictions"
        else:
            return f"Moderate restrictions - {disallow_count} disallow rules"
    
    def _generate_rate_limit_recommendation(self, severity: str) -> str:
        """Generate rate limiting recommendations based on severity"""
        recommendations = {
            "high": "Implement exponential backoff and request throttling. Consider API access.",
            "medium": "Use request delays and monitor rate limit headers.",
            "low": "Basic request spacing recommended.",
            "none": "No rate limiting detected."
        }
        return recommendations.get(severity, "Monitor for rate limiting.")

# ==============================================================================
# SCORING ENGINE & MAIN ENTRY POINT
# ==============================================================================

def calculate_automation_readiness_score(analysis_results: Dict) -> Dict[str, Any]:
    """
    Calculate comprehensive automation readiness score (0-100)
    with detailed breakdown and classification
    """
    score = 100
    deductions = []
    
    # User Agent Discrimination (Max -40)
    ua_analysis = analysis_results.get("user_agent_analysis", {})
    if ua_analysis.get("bot_blocked"):
        score -= 30
        deductions.append(("Bot UA blocked", -30))
    if ua_analysis.get("headless_blocked"):
        score -= 20
        deductions.append(("Headless blocked", -20))
    if ua_analysis.get("response_time_variance", 0) > 2.0:
        score -= 10
        deductions.append(("Response time discrimination", -10))
    
    # Bot Protection Services (Max -35)
    bot_protection = analysis_results.get("bot_protection_services", {})
    for service_id, service_data in bot_protection.items():
        risk_level = service_data.get("risk_level", "medium")
        confidence = service_data.get("confidence", "medium")
        
        if risk_level == "high" and confidence == "high":
            score -= 25
            deductions.append((f"High-risk protection: {service_id}", -25))
        elif risk_level == "high" or confidence == "high":
            score -= 15
            deductions.append((f"Medium-risk protection: {service_id}", -15))
        else:
            score -= 10
            deductions.append((f"Low-risk protection: {service_id}", -10))
    
    # Fingerprinting (Max -25)
    fingerprinting = analysis_results.get("fingerprinting_analysis", {})
    fp_summary = fingerprinting.get("summary", {})
    
    if fp_summary.get("overall_risk") == "high":
        score -= 20
        deductions.append(("Advanced fingerprinting detected", -20))
    elif fp_summary.get("overall_risk") == "medium":
        score -= 15
        deductions.append(("Moderate fingerprinting detected", -15))
    elif fp_summary.get("total_techniques", 0) > 0:
        score -= 5
        deductions.append(("Basic fingerprinting detected", -5))
    
    # Rate Limiting (Max -20)
    rate_limiting = analysis_results.get("rate_limiting_analysis", {})
    if rate_limiting.get("severity") == "high":
        score -= 15
        deductions.append(("Aggressive rate limiting", -15))
    elif rate_limiting.get("severity") == "medium":
        score -= 10
        deductions.append(("Moderate rate limiting", -10))
    elif rate_limiting.get("detected"):
        score -= 5
        deductions.append(("Basic rate limiting", -5))
    
    # Robots.txt (Max -15)
    robots_analysis = analysis_results.get("robots_txt_analysis", {})
    if robots_analysis.get("risk_level") == "high":
        score -= 10
        deductions.append(("Restrictive robots.txt", -10))
    elif robots_analysis.get("risk_level") == "medium":
        score -= 5
        deductions.append(("Moderate robots restrictions", -5))
    
    # JavaScript Challenges (Max -20)
    js_analysis = analysis_results.get("javascript_analysis", {})
    if js_analysis.get("cloudflare_challenge", {}).get("detected"):
        score -= 15
        deductions.append(("Cloudflare challenge detected", -15))
    if js_analysis.get("obfuscated_javascript", {}).get("detected"):
        score -= 10
        deductions.append(("Obfuscated JS detected", -10))
    
    # Ensure score is within bounds
    final_score = max(0, min(100, score))
    
    # Classification
    if final_score >= 80:
        classification = "READY"
        readiness_level = "Agent-friendly"
    elif final_score >= 60:
        classification = "LIMITED"
        readiness_level = "Minor friction"
    elif final_score >= 40:
        classification = "RESTRICTED" 
        readiness_level = "Significant barriers"
    else:
        classification = "BLOCKED"
        readiness_level = "High failure risk"
    
    return {
        "score": final_score,
        "classification": classification,
        "readiness_level": readiness_level,
        "breakdown": {
            "base_score": 100,
            "final_score": final_score,
            "total_deductions": 100 - final_score,
            "deduction_details": deductions
        }
    }

def check_anti_automation_readiness(url: str, timeout: int = 15) -> Dict[str, Any]:
    """
    Main entry point: Comprehensive anti-automation readiness analysis
    """
    print(f"üöÄ Starting advanced anti-automation analysis for: {url}")
    
    detector = AntiAutomationDetector(timeout=timeout)
    
    # 1. User Agent Discrimination Testing
    print("üìä Testing user agent discrimination...")
    ua_results = detector.test_user_agent_discrimination(url)
    primary_response = ua_results["raw_results"]["desktop_chrome"]
    
    # 2. Bot Protection Service Detection
    print("üõ°Ô∏è Scanning for bot protection services...")
    bot_protection = detector.detect_bot_protection_services(
        primary_response["content"], 
        primary_response["headers"]
    )
    
    # 3. Fingerprinting Analysis
    print("üîç Analyzing fingerprinting techniques...")
    fingerprinting = detector.analyze_fingerprinting(primary_response["content"])
    
    # 4. Rate Limiting Detection
    print("‚è∞ Checking for rate limiting...")
    rate_limiting = detector.detect_rate_limiting(
        primary_response["headers"],
        primary_response["status_code"]
    )
    
    # 5. Robots.txt Analysis
    print("ü§ñ Analyzing robots.txt...")
    robots_analysis = detector.analyze_robots_txt(url)
    
    # 6. JavaScript Challenge Detection
    print("üíª Detecting JavaScript challenges...")
    js_analysis = detector.analyze_javascript_challenges(
        primary_response["content"],
        primary_response["headers"]
    )
    
    # Compile comprehensive results
    analysis_results = {
        "url": url,
        "timestamp": time.time(),
        "primary_response": {
            "status_code": primary_response["status_code"],
            "response_time": primary_response["elapsed_time"],
            "content_length": primary_response["content_length"],
            "final_url": primary_response["final_url"]
        },
        "user_agent_analysis": ua_results["discrimination_analysis"],
        "bot_protection_services": bot_protection,
        "fingerprinting_analysis": fingerprinting,
        "rate_limiting_analysis": rate_limiting,
        "robots_txt_analysis": robots_analysis,
        "javascript_analysis": js_analysis,
        "raw_ua_results": {k: {
            "status": v["status_code"],
            "time": v["elapsed_time"],
            "success": v["success"]
        } for k, v in ua_results["raw_results"].items()}
    }
    
    # Calculate final score
    scoring_results = calculate_automation_readiness_score(analysis_results)
    analysis_results.update(scoring_results)
    
    # Generate recommendations
    analysis_results["recommendations"] = generate_automation_recommendations(analysis_results)
    analysis_results["issues"] = generate_issue_list(analysis_results)
    
    print(f"‚úÖ Analysis complete! Score: {scoring_results['score']}/100 - {scoring_results['classification']}")
    
    return analysis_results

def generate_automation_recommendations(analysis: Dict) -> List[str]:
    """Generate actionable recommendations based on analysis results"""
    recommendations = []
    score = analysis.get("score", 0)
    
    if score >= 80:
        recommendations.append("‚úÖ Site is automation-friendly. Standard HTTP clients should work.")
        recommendations.append("‚úÖ Minimal anti-bot measures detected. Proceed with normal automation.")
    
    elif score >= 60:
        recommendations.append("‚ö†Ô∏è Minor automation barriers detected. Consider request headers optimization.")
        recommendations.append("‚ö†Ô∏è Implement proper user-agent rotation for sustained access.")
    
    elif score >= 40:
        recommendations.append("üö® Significant automation barriers present. Headless browser recommended.")
        recommendations.append("üö® Implement advanced fingerprint rotation and request throttling.")
        recommendations.append("üö® Consider API access if available.")
    
    else:
        recommendations.append("üî¥ Heavy anti-automation measures detected. Advanced techniques required.")
        recommendations.append("üî¥ Use residential proxies and mobile device emulation.")
        recommendations.append("üî¥ Enterprise-grade bypass solutions may be necessary.")
    
    # Specific recommendations based on findings
    if analysis.get("user_agent_analysis", {}).get("bot_blocked"):
        recommendations.append("üîß Use browser-like user agents instead of bot identifiers")
    
    if analysis.get("user_agent_analysis", {}).get("headless_blocked"):
        recommendations.append("üîß Implement headless browser detection evasion")
    
    if analysis.get("bot_protection_services"):
        services = list(analysis["bot_protection_services"].keys())
        recommendations.append(f"üîß Detected protection services: {', '.join(services)}")
        for service_id, service_data in analysis["bot_protection_services"].items():
            rec = service_data.get("bypass_recommendation")
            if rec:
                recommendations.append(f"üîß {service_id}: {rec}")
    
    if analysis.get("fingerprinting_analysis", {}).get("summary", {}).get("overall_risk") in ["high", "medium"]:
        recommendations.append("üîß Implement canvas and WebGL fingerprint spoofing")
    
    if analysis.get("rate_limiting_analysis", {}).get("detected"):
        recommendations.append("üîß Implement exponential backoff and request throttling")
    
    return recommendations

# ... (Continuing the user's code right before the cutoff)

def generate_issue_list(analysis: Dict) -> List[str]:
    """Generate list of identified issues"""
    issues = []

    # User Agent Issues
    ua_analysis = analysis.get("user_agent_analysis", {})
    if ua_analysis.get("bot_blocked"):
        issues.append("Bot user agents are blocked (403/503 responses)")
    if ua_analysis.get("headless_blocked"):
        issues.append("Headless browser detection active")
    if ua_analysis.get("response_time_variance", 0) > 1.0:
        issues.append("Response time discrimination between user agents")

    # Bot Protection Issues
    bot_services = analysis.get("bot_protection_services", {})
    for service_id, service_data in bot_services.items():
        issues.append(f"Major anti-bot service detected: {service_data['service_name']} (Risk: {service_data['risk_level']})")

    # Fingerprinting Issues
    fp_summary = analysis.get("fingerprinting_analysis", {}).get("summary", {})
    if fp_summary.get("overall_risk") in ["high", "medium"]:
        issues.append(f"Extensive JS fingerprinting detected (Risk: {fp_summary['overall_risk']})")
        techniques = ', '.join(analysis['fingerprinting_analysis']['techniques_detected'].keys())
        if techniques:
            issues.append(f"Specific fingerprinting techniques: {techniques}")

    # Rate Limiting Issues
    rate_limit = analysis.get("rate_limiting_analysis", {})
    if rate_limit.get("severity") in ["high", "medium"]:
        indicators = ', '.join(rate_limit['indicators'].keys())
        issues.append(f"Active Rate Limiting detected (Severity: {rate_limit['severity']}) - Indicators: {indicators}")

    # Robots.txt Issues
    robots_analysis = analysis.get("robots_txt_analysis", {})
    if robots_analysis.get("risk_level") == "high":
        issues.append("Robots.txt contains highly restrictive disallow rules.")

    # JavaScript Challenge Issues
    js_analysis = analysis.get("javascript_analysis", {})
    if js_analysis.get("cloudflare_challenge", {}).get("detected"):
        issues.append("Cloudflare JS challenge page detected (High friction)")
    if js_analysis.get("obfuscated_javascript", {}).get("detected"):
        issues.append("Obfuscated JavaScript patterns detected")

    # Clean up and return unique issues
    return sorted(list(set(issues)))
    
# ==============================================================================
# ENHANCED VISUALIZATION MODULE (Attractive Output)
# ==============================================================================

def visualize_full_report(analysis_results: Dict):
    """
    Generates attractive, comprehensive plots and tables for the analysis results.
    """
    score = analysis_results.get("score", 0)
    classification = analysis_results.get("classification", "N/A")
    readiness_level = analysis_results.get("readiness_level", "N/A")
    url = analysis_results.get("url", "N/A")

    # ---------------------------------
    # 1. Main Score Gauge (Attractive)
    # ---------------------------------
    print("## üéØ Anti-Automation Readiness Score")
    
    # Create a DataFrame for the score breakdown
    deductions = analysis_results['breakdown']['deduction_details']
    df_deductions = pd.DataFrame(deductions, columns=['Factor', 'Deduction']).sort_values(by='Deduction')
    df_deductions['Deduction'] = df_deductions['Deduction'].abs()
    
    # Prepare data for the score gauge/pie chart
    labels = df_deductions['Factor'].tolist() + ['Final Score']
    sizes = df_deductions['Deduction'].tolist() + [score]
    
    # Define colors
    deduction_colors = sns.color_palette("Reds", n_colors=len(df_deductions))
    score_color = ['#3CB371' if score >= 60 else ('#FFA500' if score >= 40 else '#FF6347')]
    colors = list(deduction_colors) + score_color

    
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.pie(sizes, labels=labels, colors=colors, startangle=90, 
            autopct=lambda p: f'{p:.1f}%' if p > 5 else None, 
            wedgeprops={'edgecolor': 'black', 'width': 0.4})
    
    # Draw white center for donut chart
    centre_circle = plt.Circle((0,0), 0.25, color='white', fc='white', linewidth=1.25)
    plt.gca().add_artist(centre_circle)
    
    # Display score in the center
    plt.text(0, 0, f'{score:.1f}/100\n({classification})', ha='center', va='center', fontsize=18, fontweight='bold')
    plt.title(f"Overall Automation Readiness: {readiness_level}", fontsize=14)
    plt.tight_layout()
    plt.show()
    # 

    # ---------------------------------
    # 2. Key Findings & Issues
    # ---------------------------------
    print("---")
    print("## üõ°Ô∏è Key Findings Summary")
    
    # Create a DataFrame for the main issues
    df_issues = pd.DataFrame(analysis_results['issues'], columns=['Identified Issues'])
    display(df_issues.style.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#ff4d4d'), ('color', 'white')]},
        {'selector': 'td', 'props': [('border', '1px solid #ddd'), ('background-color', '#f8d7da')]}
    ]).set_caption("Critical Automation Blocking Issues"))

    # ---------------------------------
    # 3. User Agent & Response Time Analysis
    # ---------------------------------
    print("---")
    print("## üìä User Agent Discrimination Test")
    
    ua_df_data = analysis_results["raw_ua_results"]
    ua_names = list(ua_df_data.keys())
    statuses = [data['status'] for data in ua_df_data.values()]
    times = [data['time'] for data in ua_df_data.values()]
    
    df_ua = pd.DataFrame({
        'User Agent Profile': ua_names,
        'Status Code': statuses,
        'Response Time (s)': times
    })

    # Style UA table to highlight failures (4xx/5xx)
    def color_status(val):
        color = 'red' if val >= 400 or val == 0 else ('orange' if val >= 300 else 'green')
        return f'background-color: {color}; color: white'

    display(df_ua.style.applymap(color_status, subset=['Status Code']).set_caption("Multi-UA Response Comparison"))
    
    # Plot Response Times
    plt.figure(figsize=(10, 4))
    sns.barplot(x='User Agent Profile', y='Response Time (s)', data=df_ua, 
                palette=['#3CB371' if s == 200 else '#FF6347' for s in statuses])
    plt.title("Response Time & Status by User Agent")
    plt.xticks(rotation=15, ha='right')
    plt.tight_layout()
    plt.show()

    # ---------------------------------
    # 4. Bot Protection & Fingerprinting
    # ---------------------------------
    print("---")
    print("## üîç Advanced Detection & Countermeasures")

    # Bot Protection Table
    bot_data = analysis_results.get("bot_protection_services", {})
    if bot_data:
        bot_rows = []
        for svc, data in bot_data.items():
            bot_rows.append({
                "Service": data['service_name'],
                "Risk": data['risk_level'].upper(),
                "Evidence": data['evidence_count'],
                "Recommendation": data['bypass_recommendation']
            })
        df_bot = pd.DataFrame(bot_rows)
        display(df_bot.style.set_caption("Detected Anti-Bot Services"))
    else:
        print("‚úÖ No major commercial anti-bot services detected.")
        
    # Fingerprinting Chart
    fp_summary = analysis_results['fingerprinting_analysis']['summary']
    if fp_summary['total_techniques'] > 0:
        fp_techniques = analysis_results['fingerprinting_analysis']['techniques_detected']
        fp_df = pd.DataFrame([
            (k.capitalize(), v['pattern_count'], v['risk_level']) 
            for k, v in fp_techniques.items()
        ], columns=['Technique', 'Pattern Count', 'Risk Level'])

        plt.figure(figsize=(10, 5))
        sns.barplot(x='Technique', y='Pattern Count', data=fp_df, 
                    palette=['#ff4d4d' if r == 'high' else ('#FFA500' if r == 'medium' else '#3CB371') 
                             for r in fp_df['Risk Level']])
        plt.title(f"Detected JS Fingerprinting Techniques (Overall Risk: {fp_summary['overall_risk'].upper()})")
        plt.xticks(rotation=30, ha='right')
        plt.tight_layout()
        plt.show()
    # 

    # ---------------------------------
    # 5. Recommendations
    # ---------------------------------
    print("---")
    print("## ‚úÖ Actionable Recommendations")
    df_reco = pd.DataFrame(analysis_results['recommendations'], columns=['Recommendation'])
    display(df_reco.style.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#3CB371'), ('color', 'white')]},
        {'selector': 'td', 'props': [('border', '1px solid #ddd')]}
    ]).set_caption("Strategy for Automation Success"))


# ==============================================================================
# FINAL EXECUTION EXAMPLE
# ==============================================================================

# Example usage (Replace with your actual target URL)
# The URL below is a dummy, please use a test site or a site you have permission to test.
# target_url = "http://httpbin.org/get" 
# analysis_result = check_anti_automation_readiness(target_url)

# Example output from a theoretical test to display the report:
# (Uncomment the lines above and comment the lines below to run a real test)

if 'analysis_result' not in locals():
    # Placeholder to demonstrate the visualization using the scoring logic's internal dummy data
    print("\n--- DEMONSTRATION MODE ---")
    
    # We'll re-run the core logic on a simple URL and use its results for visualization
    # Note: Running this against a real site requires the 'requests' library and network access.
    
    # Since I cannot run network requests, I will construct a plausible mock result
    mock_analysis_result = {
        'url': 'https://www.example-high-protection.com',
        'score': 35,
        'classification': 'BLOCKED',
        'readiness_level': 'High failure risk',
        'breakdown': {
            'deduction_details': [
                ('Bot UA blocked', -30),
                ('Headless blocked', -20),
                ('High-risk protection: cloudflare', -25),
                ('Advanced fingerprinting detected', -20),
                ('Aggressive rate limiting', -15)
            ]
        },
        'user_agent_analysis': {'bot_blocked': True, 'headless_blocked': True, 'response_time_variance': 3.5},
        'bot_protection_services': {
            'cloudflare': {'service_name': 'Cloudflare Bot Management', 'risk_level': 'high', 'evidence_count': 5, 'bypass_recommendation': 'Use Cloudflare API tokens or residential proxies'}
        },
        'fingerprinting_analysis': {
            'summary': {'overall_risk': 'high', 'total_techniques': 4},
            'techniques_detected': {'canvas': {'pattern_count': 3, 'risk_level': 'high'}, 'webgl': {'pattern_count': 2, 'risk_level': 'high'}, 'hardware': {'pattern_count': 5, 'risk_level': 'medium'}}
        },
        'rate_limiting_analysis': {'detected': True, 'severity': 'high', 'indicators': {'x-ratelimit-limit': {'value': '60'}}},
        'robots_txt_analysis': {'risk_level': 'low'},
        'javascript_analysis': {'cloudflare_challenge': {'detected': True}},
        'raw_ua_results': {
            'desktop_chrome': {'status': 200, 'time': 1.2, 'success': True}, 
            'python_bot': {'status': 403, 'time': 0.8, 'success': False}, 
            'headless_chrome': {'status': 403, 'time': 1.5, 'success': False}
        },
        'issues': generate_issue_list({
            'user_agent_analysis': {'bot_blocked': True, 'headless_blocked': True, 'response_time_variance': 3.5},
            'bot_protection_services': {'cloudflare': {'service_name': 'Cloudflare Bot Management', 'risk_level': 'high', 'evidence_count': 5, 'bypass_recommendation': '...'}},
            'fingerprinting_analysis': {'summary': {'overall_risk': 'high'}, 'techniques_detected': {'canvas': {'pattern_count': 3, 'risk_level': 'high'}}},
            'rate_limiting_analysis': {'detected': True, 'severity': 'high', 'indicators': {'x-ratelimit-limit': {'value': '60'}}},
            'robots_txt_analysis': {'risk_level': 'low'},
            'javascript_analysis': {'cloudflare_challenge': {'detected': True}}
        })[0:5], # Keep list short for display
        'recommendations': generate_automation_recommendations(
            {'score': 35, 'user_agent_analysis': {'bot_blocked': True}, 'bot_protection_services': {'cloudflare': {'bypass_recommendation': 'Use API'}}, 'fingerprinting_analysis': {'summary': {'overall_risk': 'high'}}, 'rate_limiting_analysis': {'detected': True}}
        )
    }
    
    visualize_full_report(mock_analysis_result)
````

````python
"""
üöÄ ENHANCED ANTI-AUTOMATION READINESS MODULE
============================================
Bridge AI's ARI Framework - Production-Grade Bot Detection Analysis

This module evaluates website readiness for autonomous AI agents by detecting:
- Multi-layer bot protection mechanisms
- JavaScript challenges and CAPTCHAs  
- Rate limiting and fingerprinting
- User-agent discrimination
- Dynamic protection systems

IMPROVEMENTS OVER LEGACY VERSION:
‚úÖ Multi-User-Agent testing (6 profiles)
‚úÖ Advanced bot protection service detection (8 services)
‚úÖ Comprehensive fingerprinting analysis (12+ patterns)
‚úÖ Rate limiting detection with severity scoring
‚úÖ Intelligent robots.txt parsing
‚úÖ JavaScript challenge detection
‚úÖ Granular 0-100 scoring system
‚úÖ Parallel execution for performance
‚úÖ Production-ready error handling
"""

import requests
import re
import time
import json
from typing import Dict, List, Tuple, Any
from urllib.parse import urlparse
from concurrent.futures import ThreadPoolExecutor, as_completed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

# ============================================================================
# ENHANCED CONFIGURATION
# ============================================================================

# Comprehensive User Agent Profiles for Discrimination Testing
USER_AGENT_PROFILES = {
    "desktop_chrome": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "desktop_firefox": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:121.0) Gecko/20100101 Firefox/121.0", 
    "mobile_ios": "Mozilla/5.0 (iPhone; CPU iPhone OS 17_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.2 Mobile/15E148 Safari/604.1",
    "headless_chrome": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) HeadlessChrome/120.0.0.0 Safari/537.36",
    "python_bot": "python-requests/2.31.0",
    "legacy_browser": "Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko"
}

# Advanced Bot Protection Service Signatures
BOT_PROTECTION_SIGNATURES = {
    "cloudflare": {
        "signatures": ["cloudflare", "cf-ray", "__cf_bm", "cf_clearance", "checking your browser", "jschl-answer"],
        "headers": ["cf-ray", "server", "cf-cache-status"],
        "risk_level": "high"
    },
    "perimeterx": {
        "signatures": ["perimeterx", "_px", "px-captcha", "pxinit", "humansecurity"],
        "headers": ["x-px", "server"],
        "risk_level": "high" 
    },
    "datadome": {
        "signatures": ["datadome", "ddg.js", "ddg_bot", "ddg_mobile"],
        "headers": ["x-datadome"],
        "risk_level": "high"
    },
    "akamai": {
        "signatures": ["akamai", "ak_bm", "_abck", "akacd"],
        "headers": ["x-akamai-transformed", "server"],
        "risk_level": "medium"
    },
    "recaptcha": {
        "signatures": ["recaptcha", "grecaptcha", "g-recaptcha", "recaptcha/api"],
        "headers": [],
        "risk_level": "medium"
    },
    "hcaptcha": {
        "signatures": ["hcaptcha", "h-captcha", "hcaptcha.com"],
        "headers": [],
        "risk_level": "medium"
    },
    "arkose": {
        "signatures": ["arkoselabs", "funcaptcha", "arkose"],
        "headers": [],
        "risk_level": "high"
    },
    "shape": {
        "signatures": ["shapesecurity", "shape_rc", "ss_rc"],
        "headers": ["x-shape-request-id"],
        "risk_level": "high"
    }
}

# Comprehensive Fingerprinting Detection Patterns
FINGERPRINT_PATTERNS = [
    # Canvas fingerprinting
    r"canvas\.(getContext|toDataURL|toBlob)",
    r"fillText.*fingerprint", 
    r"getImageData",
    
    # WebGL fingerprinting
    r"webgl",
    r"getParameter.*webgl",
    r"renderer.*webgl",
    
    # Audio fingerprinting
    r"audiocontext",
    r"createanalyser",
    r"getchanneldata",
    
    # Hardware detection
    r"hardwareconcurrency",
    r"devicememory", 
    r"maxTouchPoints",
    
    # Screen properties
    r"screen\.(width|height|avail)",
    r"pixelDepth",
    r"colorDepth",
    
    # Plugin enumeration
    r"navigator\.plugins",
    r"navigator\.mimeTypes",
    
    # WebRTC detection
    r"rtcpeerconnection",
    r"createDataChannel",
    
    # Font detection
    r"getComputedStyle",
    r"fonts\.check",
    
    # Battery API
    r"getBattery",
    r"battery.*level",
    
    # Obfuscation patterns
    r"eval.*function",
    r"escape.*unescape",
    r"fromCharCode"
]

# Rate Limiting Detection Headers
RATE_LIMIT_HEADERS = [
    "x-ratelimit-limit", "x-ratelimit-remaining", "x-ratelimit-reset",
    "retry-after", "x-rate-limit-limit", "x-rate-limit-remaining"
]

# ============================================================================
# CORE DETECTION FUNCTIONS
# ============================================================================

def fetch_with_headers(url: str, user_agent: str, timeout: int = 10) -> Dict[str, Any]:
    """
    Enhanced URL fetching with comprehensive metrics and error handling
    
    Args:
        url: Target URL to fetch
        user_agent: User agent string to use
        timeout: Request timeout in seconds
        
    Returns:
        Dict containing response data, timing, and error information
    """
    try:
        start_time = time.time()
        response = requests.get(
            url,
            headers={
                "User-Agent": user_agent,
                "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
                "Accept-Language": "en-US,en;q=0.9",
                "Accept-Encoding": "gzip, deflate, br",
                "DNT": "1",
                "Connection": "keep-alive",
                "Upgrade-Insecure-Requests": "1"
            },
            timeout=timeout,
            allow_redirects=True
        )
        elapsed = time.time() - start_time
        
        return {
            "status": response.status_code,
            "headers": dict(response.headers),
            "content": response.text,
            "elapsed": elapsed,
            "url": response.url,
            "redirected": response.url != url,
            "content_length": len(response.text),
            "success": True
        }
    except requests.exceptions.Timeout:
        return {
            "status": 0, 
            "error": "timeout", 
            "headers": {}, 
            "content": "", 
            "elapsed": timeout,
            "success": False
        }
    except Exception as e:
        return {
            "status": 0, 
            "error": str(e), 
            "headers": {}, 
            "content": "", 
            "elapsed": 0,
            "success": False
        }


def analyze_user_agent_discrimination(url: str) -> Dict[str, Any]:
    """
    Test website with multiple user-agents to detect blocking patterns
    
    Uses parallel execution for performance. Tests 6 different user agents
    to identify discrimination against bots, headless browsers, etc.
    
    Returns:
        Dict with discrimination analysis and raw results
    """
    print("  ‚Üí Testing 6 user agents in parallel...")
    results = {}
    
    # Execute all user agent tests in parallel
    with ThreadPoolExecutor(max_workers=4) as executor:
        future_to_ua = {
            executor.submit(fetch_with_headers, url, ua_string): ua_name 
            for ua_name, ua_string in USER_AGENT_PROFILES.items()
        }
        
        for future in as_completed(future_to_ua):
            ua_name = future_to_ua[future]
            results[ua_name] = future.result()
    
    # Analyze discrimination patterns
    desktop_ok = results["desktop_chrome"]["status"] == 200
    bot_blocked = results["python_bot"]["status"] in [403, 429, 503]
    headless_blocked = results["headless_chrome"]["status"] in [403, 503]
    mobile_diff = results["mobile_ios"]["status"] != results["desktop_chrome"]["status"]
    legacy_diff = results["legacy_browser"]["status"] != results["desktop_chrome"]["status"]
    
    # Calculate response time variance
    times = [data["elapsed"] for data in results.values() if data["success"]]
    time_variance = max(times) - min(times) if len(times) > 1 else 0
    
    return {
        "raw_results": results,
        "discrimination_analysis": {
            "bot_blocked": bot_blocked and desktop_ok,
            "headless_blocked": headless_blocked and desktop_ok, 
            "mobile_different": mobile_diff,
            "legacy_different": legacy_diff,
            "response_time_variance": time_variance
        }
    }


def detect_bot_protection_services(content: str, headers: Dict) -> Dict[str, Any]:
    """
    Identify known bot protection services with confidence scoring
    
    Scans both content and headers for signatures of major bot protection
    services like Cloudflare, PerimeterX, DataDome, etc.
    
    Returns:
        Dict of detected services with evidence and confidence levels
    """
    detected_services = {}
    content_lower = content.lower()
    
    for service_id, service_config in BOT_PROTECTION_SIGNATURES.items():
        content_matches = []
        header_matches = []
        
        # Check content signatures
        for signature in service_config["signatures"]:
            if signature.lower() in content_lower:
                content_matches.append(signature)
        
        # Check header signatures  
        for header_name, header_value in headers.items():
            header_str = f"{header_name}: {header_value}".lower()
            for signature in service_config["signatures"] + service_config["headers"]:
                if signature.lower() in header_str:
                    header_matches.append(f"{header_name}: {signature}")
        
        # Only report if we found evidence
        if content_matches or header_matches:
            confidence = "high" if len(content_matches) > 2 or len(header_matches) > 1 else "medium"
            
            detected_services[service_id] = {
                "service_name": service_id.title(),
                "confidence": confidence,
                "risk_level": service_config["risk_level"],
                "content_matches": content_matches[:3],  # Limit output
                "header_matches": header_matches[:3],
                "evidence_count": len(content_matches) + len(header_matches)
            }
    
    return detected_services


def analyze_fingerprinting(content: str) -> Dict[str, Any]:
    """
    Detect advanced browser fingerprinting techniques
    
    Identifies canvas, WebGL, audio, hardware, and other fingerprinting
    methods. Includes obfuscated script detection via entropy analysis.
    
    Returns:
        Comprehensive fingerprinting analysis with risk assessment
    """
    detected_patterns = []
    content_lower = content.lower()
    
    # Pattern-based detection
    for pattern in FINGERPRINT_PATTERNS:
        matches = re.findall(pattern, content_lower, re.IGNORECASE)
        if matches:
            detected_patterns.append({
                "pattern": pattern,
                "count": len(matches),
                "examples": matches[:2]  # Sample matches
            })
    
    # Obfuscated script detection (entropy analysis)
    script_blocks = re.findall(r'<script[^>]*>(.*?)</script>', content, re.DOTALL | re.IGNORECASE)
    high_entropy_scripts = 0
    
    for script in script_blocks:
        if len(script) > 500:  # Only check substantial scripts
            # Simple entropy check - look for long strings with few spaces
            words = script.split()
            if len(words) > 10:  # Enough words for meaningful analysis
                avg_word_length = sum(len(word) for word in words) / len(words)
                if avg_word_length > 15:  # Likely obfuscated
                    high_entropy_scripts += 1
    
    # Risk assessment
    pattern_count = len(detected_patterns)
    if pattern_count > 5 or high_entropy_scripts > 2:
        risk_level = "high"
    elif pattern_count > 2 or high_entropy_scripts > 0:
        risk_level = "medium" 
    else:
        risk_level = "low"
    
    return {
        "patterns_detected": detected_patterns,
        "pattern_count": pattern_count,
        "obfuscated_scripts": high_entropy_scripts,
        "risk_level": risk_level,
        "techniques_identified": list(set([p["pattern"].split('.')[0] for p in detected_patterns]))
    }


def analyze_robots_txt(url: str) -> Dict[str, Any]:
    """
    Intelligent robots.txt analysis with context-aware scoring
    
    Parses robots.txt for meaningful restrictions beyond simple "Disallow: /"
    checks. Analyzes user-agent specific rules and crawl delays.
    
    Returns:
        Detailed robots.txt analysis with restriction assessment
    """
    parsed = urlparse(url)
    robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
    
    try:
        response = requests.get(robots_url, timeout=8)
        if response.status_code != 200:
            return {
                "exists": False, 
                "accessible": False,
                "restrictive": False,
                "risk_level": "low"
            }
        
        content = response.text
        lines = [line.strip() for line in content.split('\n') if line.strip() and not line.startswith('#')]
        
        # Parse restrictions
        disallow_all = any("disallow: /" in line.lower() for line in lines)
        disallow_patterns = [line.split('disallow:')[1].strip() for line in lines if 'disallow:' in line.lower()]
        allow_patterns = [line.split('allow:')[1].strip() for line in lines if 'allow:' in line.lower()]
        has_crawl_delay = any('crawl-delay:' in line.lower() for line in lines)
        user_agent_specific = any('user-agent:' in line.lower() for line in lines)
        
        # Risk assessment
        if disallow_all:
            risk_level = "high"
        elif len(disallow_patterns) > 8:
            risk_level = "medium"
        elif len(disallow_patterns) > 0:
            risk_level = "low"
        else:
            risk_level = "none"
        
        return {
            "exists": True,
            "accessible": True,
            "disallow_all": disallow_all,
            "restrictive": len(disallow_patterns) > 5,
            "user_agent_specific": user_agent_specific,
            "crawl_delay": has_crawl_delay,
            "disallow_count": len(disallow_patterns),
            "allow_count": len(allow_patterns),
            "risk_level": risk_level,
            "analysis": "Blocks entire site" if disallow_all else 
                       f"Has {len(disallow_patterns)} disallow rules" if disallow_patterns else "Permissive"
        }
        
    except Exception as e:
        return {
            "exists": False,
            "accessible": False, 
            "error": str(e),
            "restrictive": False,
            "risk_level": "low"
        }


def detect_rate_limiting(ua_results: Dict) -> Dict[str, Any]:
    """
    Comprehensive rate limiting detection from headers and status codes
    
    Analyzes responses for rate limiting indicators including:
    - HTTP 429 status codes
    - X-RateLimit-* headers
    - Retry-After headers
    - Response time anomalies
    
    Returns:
        Rate limiting analysis with severity assessment
    """
    rate_detected = False
    indicators = []
    
    for ua_name, data in ua_results.items():
        status = data.get("status", 0)
        headers = data.get("headers", {})
        
        # Status code detection
        if status == 429:
            rate_detected = True
            indicators.append(f"{ua_name}: HTTP 429 (Too Many Requests)")
        
        # Header-based detection
        for header_name, header_value in headers.items():
            if any(rate_header in header_name.lower() for rate_header in RATE_LIMIT_HEADERS):
                rate_detected = True
                indicators.append(f"{ua_name}: {header_name} = {header_value}")
                break
    
    # Severity classification
    if any("HTTP 429" in ind for ind in indicators):
        severity = "high"
    elif len(indicators) > 2:
        severity = "medium" 
    elif rate_detected:
        severity = "low"
    else:
        severity = "none"
    
    return {
        "detected": rate_detected,
        "severity": severity,
        "indicators": indicators,
        "recommendation": "Implement exponential backoff" if severity == "high" else
                         "Use request throttling" if severity == "medium" else
                         "Monitor for rate limits" if severity == "low" else
                         "No rate limiting detected"
    }


def analyze_javascript_challenges(content: str, headers: Dict) -> Dict[str, Any]:
    """
    Detect JavaScript-based challenges and protections
    
    Identifies Cloudflare challenges, obfuscated JavaScript, and other
    client-side protection mechanisms that may block automation.
    
    Returns:
        JavaScript challenge analysis
    """
    js_analysis = {}
    content_lower = content.lower()
    
    # Cloudflare challenge detection
    cloudflare_indicators = [
        "checking your browser", "ddos protection", "please wait", 
        "cloudflare ray id", "jschl-answer", "challenge-platform"
    ]
    cf_matches = [ind for ind in cloudflare_indicators if ind in content_lower]
    
    if cf_matches:
        js_analysis["cloudflare_challenge"] = {
            "detected": True,
            "indicators": cf_matches,
            "risk_level": "high"
        }
    
    # Obfuscated JavaScript detection
    obfuscation_patterns = [
        r"eval\(.*atob", r"document\.write.*fromCharCode", 
        r"setInterval.*decode", r"function.*constructor"
    ]
    obfuscated_matches = []
    for pattern in obfuscation_patterns:
        matches = re.findall(pattern, content_lower)
        obfuscated_matches.extend(matches)
    
    if obfuscated_matches:
        js_analysis["obfuscated_javascript"] = {
            "detected": True, 
            "pattern_count": len(obfuscated_matches),
            "risk_level": "medium"
        }
    
    # CAPTCHA detection
    captcha_indicators = ["captcha", "recaptcha", "hcaptcha", "funcaptcha"]
    captcha_matches = [ind for ind in captcha_indicators if ind in content_lower]
    
    if captcha_matches:
        js_analysis["captcha"] = {
            "detected": True,
            "types": captcha_matches,
            "risk_level": "medium"
        }
    
    return js_analysis


def calculate_readiness_score(analysis: Dict) -> Tuple[int, str, str]:
    """
    Calculate comprehensive automation readiness score (0-100)
    
    Scoring Breakdown:
    - User Agent Discrimination: -40 points max
    - Bot Protection Services: -35 points max  
    - Fingerprinting: -25 points max
    - Rate Limiting: -20 points max
    - JavaScript Challenges: -20 points max
    - Robots.txt: -15 points max
    
    Returns:
        Tuple of (score, classification, readiness_level)
    """
    score = 100
    
    # User Agent Discrimination penalties
    ua_analysis = analysis["ua_discrimination"]
    if ua_analysis["bot_blocked"]:
        score -= 30
    if ua_analysis["headless_blocked"]:
        score -= 20
    if ua_analysis["response_time_variance"] > 2.0:
        score -= 10
    
    # Bot Protection Services penalties
    bot_services = analysis["bot_protection"]
    for service_id, service_data in bot_services.items():
        if service_data["risk_level"] == "high" and service_data["confidence"] == "high":
            score -= 25
        elif service_data["risk_level"] == "high" or service_data["confidence"] == "high":
            score -= 15
        else:
            score -= 10
    
    # Fingerprinting penalties
    fingerprinting = analysis["fingerprinting"]
    if fingerprinting["risk_level"] == "high":
        score -= 20
    elif fingerprinting["risk_level"] == "medium":
        score -= 10
    
    # Rate Limiting penalties
    rate_limiting = analysis["rate_limiting"]
    if rate_limiting["severity"] == "high":
        score -= 15
    elif rate_limiting["severity"] == "medium":
        score -= 10
    elif rate_limiting["detected"]:
        score -= 5
    
    # JavaScript Challenges penalties
    js_analysis = analysis["javascript_analysis"]
    if js_analysis.get("cloudflare_challenge", {}).get("detected"):
        score -= 15
    if js_analysis.get("obfuscated_javascript", {}).get("detected"):
        score -= 10
    
    # Robots.txt penalties
    robots_txt = analysis["robots_txt"]
    if robots_txt.get("risk_level") == "high":
        score -= 10
    elif robots_txt.get("risk_level") == "medium":
        score -= 5
    
    # Ensure score is within bounds
    final_score = max(0, min(100, score))
    
    # Classification and readiness level
    if final_score >= 80:
        classification = "READY"
        readiness_level = "Agent-friendly"
    elif final_score >= 60:
        classification = "LIMITED" 
        readiness_level = "Minor friction"
    elif final_score >= 40:
        classification = "RESTRICTED"
        readiness_level = "Significant barriers"
    else:
        classification = "BLOCKED"
        readiness_level = "High failure risk"
    
    return final_score, classification, readiness_level


def generate_recommendations(analysis: Dict) -> Tuple[List[str], List[str]]:
    """
    Generate actionable issues and recommendations based on analysis
    
    Returns:
        Tuple of (issues_list, recommendations_list)
    """
    issues = []
    recommendations = []
    
    # User Agent issues
    ua_analysis = analysis["ua_discrimination"]
    if ua_analysis["bot_blocked"]:
        issues.append("Website blocks bot user-agents (403/503 responses)")
        recommendations.append("Allow generic bot UAs or provide developer API tokens")
    
    if ua_analysis["headless_blocked"]:
        issues.append("Headless browser detection active")
        recommendations.append("Relax TLS/browser fingerprint checks for legitimate automation")
    
    # Bot Protection issues
    bot_services = analysis["bot_protection"]
    for service_id, service_data in bot_services.items():
        issues.append(f"{service_data['service_name']} bot protection detected ({service_data['confidence']} confidence)")
        recommendations.append(f"Consider API access or specialized bypass for {service_data['service_name']}")
    
    # Fingerprinting issues
    fingerprinting = analysis["fingerprinting"]
    if fingerprinting["risk_level"] in ["high", "medium"]:
        issues.append(f"Advanced fingerprinting detected ({fingerprinting['pattern_count']} techniques)")
        recommendations.append("Reduce fingerprinting for legitimate automation use cases")
    
    # Rate Limiting issues
    rate_limiting = analysis["rate_limiting"]
    if rate_limiting["detected"]:
        issues.append(f"Rate limiting detected ({rate_limiting['severity']} severity)")
        recommendations.append(rate_limiting["recommendation"])
    
    # JavaScript Challenges issues
    js_analysis = analysis["javascript_analysis"]
    if js_analysis.get("cloudflare_challenge", {}).get("detected"):
        issues.append("Cloudflare JavaScript challenge protection active")
        recommendations.append("Use headless browsers or Cloudflare bypass services")
    
    if js_analysis.get("obfuscated_javascript", {}).get("detected"):
        issues.append("Obfuscated JavaScript detected (anti-debugging)")
        recommendations.append("Implement JavaScript execution analysis")
    
    # Robots.txt issues
    robots_txt = analysis["robots_txt"]
    if robots_txt.get("disallow_all"):
        issues.append("robots.txt blocks all crawling (Disallow: /)")
        recommendations.append("Update robots.txt to allow general discovery")
    elif robots_txt.get("restrictive"):
        issues.append("Restrictive robots.txt with many disallow rules")
        recommendations.append("Consider relaxing robots.txt for API discovery")
    
    return issues, recommendations


# ============================================================================
# MAIN ENTRY POINT - ENHANCED VERSION
# ============================================================================

def check_anti_automation_readiness_enhanced(url: str) -> Dict[str, Any]:
    """
    üöÄ COMPREHENSIVE ANTI-AUTOMATION READINESS ANALYSIS
    
    Enhanced version with multi-layer detection, parallel execution, and
    detailed scoring. Replaces the legacy check_anti_automation_absence function.
    
    Args:
        url: Target website URL to analyze
        
    Returns:
        Complete analysis report with score, classification, issues, 
        recommendations, and detailed technical analysis
    """
    print(f"üöÄ Starting enhanced anti-automation analysis for: {url}")
    start_time = time.time()
    
    try:
        # Step 1: Multi-User-Agent Discrimination Testing
        print("üìä Testing user agent discrimination...")
        ua_discrimination = analyze_user_agent_discrimination(url)
        
        # Get primary response for further analysis
        desktop_result = ua_discrimination["raw_results"]["desktop_chrome"]
        content = desktop_result.get("content", "")
        headers = desktop_result.get("headers", {})
        
        # Step 2: Bot Protection Service Detection
        print("üõ°Ô∏è Scanning for bot protection services...")
        bot_protection = detect_bot_protection_services(content, headers)
        
        # Step 3: Fingerprinting Analysis
        print("üîç Analyzing fingerprinting techniques...")
        fingerprinting = analyze_fingerprinting(content)
        
        # Step 4: Rate Limiting Detection
        print("‚è∞ Checking for rate limiting...")
        rate_limiting = detect_rate_limiting(ua_discrimination["raw_results"])
        
        # Step 5: Robots.txt Analysis
        print("ü§ñ Analyzing robots.txt...")
        robots_txt = analyze_robots_txt(url)
        
        # Step 6: JavaScript Challenge Detection
        print("üíª Detecting JavaScript challenges...")
        javascript_analysis = analyze_javascript_challenges(content, headers)
        
        # Compile comprehensive analysis
        analysis_results = {
            "ua_discrimination": ua_discrimination["discrimination_analysis"],
            "bot_protection": bot_protection,
            "fingerprinting": fingerprinting,
            "rate_limiting": rate_limiting,
            "robots_txt": robots_txt,
            "javascript_analysis": javascript_analysis
        }
        
        # Step 7: Calculate Score and Classification
        print("üéØ Calculating readiness score...")
        score, classification, readiness_level = calculate_readiness_score(analysis_results)
        
        # Step 8: Generate Issues and Recommendations
        print("üí° Generating recommendations...")
        issues, recommendations = generate_recommendations(analysis_results)
        
        # Calculate execution time
        execution_time = time.time() - start_time
        
        # Final comprehensive result
        result = {
            "url": url,
            "score": score,
            "classification": classification,
            "readiness_level": readiness_level,
            "issues": issues,
            "recommendations": recommendations,
            "detailed_analysis": analysis_results,
            "execution_time": execution_time,
            "timestamp": time.strftime("%Y-%m-%d %H:%M:%S"),
            
            # Additional metadata for visualization
            "ua_analysis": {
                ua_name: {
                    "status": data["status"],
                    "headers": data["headers"],
                    "elapsed": data["elapsed"]
                }
                for ua_name, data in ua_discrimination["raw_results"].items()
            },
            "js_checks": {
                "js_challenge": javascript_analysis.get("cloudflare_challenge", {}).get("detected", False),
                "captcha_detected": javascript_analysis.get("captcha", {}).get("detected", False)
            }
        }
        
        print(f"‚úÖ Analysis complete! Score: {score}/100 - {classification}")
        print(f"‚è±Ô∏è  Execution time: {execution_time:.2f}s")
        
        return result
        
    except Exception as e:
        print(f"‚ùå Analysis failed: {str(e)}")
        return {
            "url": url,
            "score": 0,
            "classification": "ERROR",
            "readiness_level": "Analysis failed",
            "issues": [f"Analysis error: {str(e)}"],
            "recommendations": ["Check URL accessibility and try again"],
            "error": str(e),
            "execution_time": time.time() - start_time,
            "timestamp": time.strftime("%Y-%m-%d %H:%M:%S")
        }


# ============================================================================
# LEGACY COMPATIBILITY FUNCTION
# ============================================================================

def check_anti_automation_absence(url: str, html: str = None) -> Tuple[int, List[str], List[str]]:
    """
    üî¥ LEGACY FUNCTION - Maintained for backward compatibility
    
    Original anti-automation check with 0-5 scoring. This is kept to ensure
    existing code continues to work while we transition to the enhanced version.
    
    Args:
        url: Target website URL
        html: Optional HTML content (for compatibility)
        
    Returns:
        Tuple of (score, issues, recommendations)
    """
    score = 0
    issues = []
    recommendations = []

    try:
        # 1. Check for bot-blocking headers
        response = requests.get(url, timeout=10)
        headers = response.headers

        if "x-robots-tag" not in headers or "noindex" not in headers.get("x-robots-tag", "").lower():
            score += 1
        else:
            issues.append("Site blocks bots via X-Robots-Tag header.")
            recommendations.append("Remove aggressive noindex/nofollow unless essential.")

        # 2. Check for CAPTCHAs on homepage
        if "captcha" in response.text.lower():
            issues.append("Potential CAPTCHA challenge found on homepage.")
            recommendations.append("Use CAPTCHAs only on sensitive actions like signups or payments.")
        else:
            score += 2

        # 3. robots.txt check
        parsed = requests.utils.urlparse(url)
        robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
        try:
            r_txt = requests.get(robots_url, timeout=5)
            if "disallow: /" not in r_txt.text.lower():
                score += 1
            else:
                issues.append("robots.txt blocks general crawling.")
                recommendations.append("Update robots.txt to allow general crawling.")
        except:
            issues.append("robots.txt not found or unreachable.")
            recommendations.append("Ensure robots.txt is accessible and well-configured.")

        # 4. Check for fingerprinting scripts (basic heuristic)
        if any(s in response.text.lower() for s in ["fingerprintjs", "navigator.plugins", "navigator.hardwareconcurrency"]):
            issues.append("Potential fingerprinting scripts detected.")
            recommendations.append("Avoid aggressive fingerprinting that blocks automation.")
        else:
            score += 1

        return min(score, 5), issues, recommendations

    except Exception as e:
        issues.append("Anti-automation check failed.")
        recommendations.append(str(e))
        return 0, issues, recommendations


# ============================================================================
# VISUALIZATION FUNCTION (FIXED VERSION)
# ============================================================================

def visualize_anti_automation_comparison(old_result: Dict, new_result: Dict):
    """
    üé® COMPREHENSIVE VISUAL COMPARISON: OLD vs ENHANCED
    
    Fixed visualization function that handles the new result structure
    and provides detailed comparison analytics.
    
    Args:
        old_result: Result from legacy check_anti_automation_absence
        new_result: Result from enhanced check_anti_automation_readiness_enhanced
    """
    print("\n" + "="*90)
    print("üöÄ ANTI-AUTOMATION READINESS COMPARISON ANALYTICS")
    print("="*90)
    
    # Handle score conversion (old: 0-5, new: 0-100)
    old_score = old_result.get("score", 0)
    old_normalized = old_score * 20  # Convert to 0-100 scale for comparison
    new_score = new_result.get("score", 0)
    
    # 1. EXECUTIVE SUMMARY
    print("\nüìä EXECUTIVE SUMMARY")
    print("-" * 50)
    print(f"üìç URL: {new_result.get('url', 'N/A')}")
    print(f"üïí Analysis Time: {new_result.get('timestamp', 'N/A')}")
    print()
    print(f"üî¥ LEGACY SCRIPT: {old_score}/5 ‚Üí {old_normalized}/100")
    print(f"üü¢ ENHANCED SCRIPT: {new_score}/100")
    print(f"üìà IMPROVEMENT: {new_score - old_normalized:+.1f} points")
    print(f"üéØ CLASSIFICATION: {old_result.get('readiness', 'N/A')} ‚Üí {new_result.get('classification', 'N/A')}")
    print()
    print(f"‚ö†Ô∏è  Issues Detected: {len(old_result.get('issues', []))} ‚Üí {len(new_result.get('issues', []))}")
    print(f"üí° Recommendations: {len(old_result.get('recommendations', []))} ‚Üí {len(new_result.get('recommendations', []))}")
    print(f"‚è±Ô∏è  Execution Time: N/A ‚Üí {new_result.get('execution_time', 0):.2f}s")
    
    # 2. SCORE COMPARISON CHART
    plt.figure(figsize=(10, 6))
    
    # Main scores
    plt.subplot(1, 2, 1)
    versions = ['Legacy\n(0-5 scale)', 'Enhanced\n(0-100 scale)']
    scores = [old_score, new_score]
    colors = ['#ff6b6b', '#6bcf7f']
    
    bars = plt.bar(versions, scores, color=colors, edgecolor='black', alpha=0.8)
    plt.ylabel('Score', fontweight='bold')
    plt.title('Score Comparison', fontweight='bold', fontsize=14)
    
    # Add value labels
    for bar, score in zip(bars, scores):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
                f'{score}', ha='center', va='bottom', fontweight='bold')
    
    # Normalized comparison
    plt.subplot(1, 2, 2)
    normalized_scores = [old_normalized, new_score]
    bars = plt.bar(['Legacy\n(Normalized)', 'Enhanced'], normalized_scores, 
                   color=['#ff9e6b', '#6bcf7f'], edgecolor='black', alpha=0.8)
    plt.ylabel('Score (0-100 scale)', fontweight='bold')
    plt.title('Normalized Comparison', fontweight='bold', fontsize=14)
    plt.ylim(0, 110)
    
    for bar, score in zip(bars, normalized_scores):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2, 
                f'{score:.0f}', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    # 3. DETECTION CAPABILITIES COMPARISON
    print("\nüîç DETECTION CAPABILITIES COMPARISON")
    print("-" * 50)
    
    capabilities = [
        ("User Agent Testing", "1 UA", f"{len(USER_AGENT_PROFILES)} UAs"),
        ("Bot Protection Services", "0", f"{len(new_result.get('detailed_analysis', {}).get('bot_protection', {}))} detected"),
        ("Fingerprinting Patterns", "3 basic", f"{new_result.get('detailed_analysis', {}).get('fingerprinting', {}).get('pattern_count', 0)} advanced"),
        ("Rate Limiting", "‚ùå Not detected", "‚úÖ Header + status analysis"),
        ("JavaScript Challenges", "‚ùå Not detected", "‚úÖ Cloudflare + obfuscation"),
        ("Robots.txt Analysis", "Basic", "Intelligent parsing"),
        ("Execution", "Sequential", "Parallel")
    ]
    
    for capability, old_status, new_status in capabilities:
        print(f"   {capability:<25} {old_status:<20} ‚Üí {new_status}")
    
    # 4. ISSUES COMPARISON TABLE
    print("\n‚ö†Ô∏è  ISSUES COMPARISON")
    print("-" * 50)
    
    max_issues = max(len(old_result.get('issues', [])), len(new_result.get('issues', [])), 1)
    old_issues = old_result.get('issues', []) + [''] * (max_issues - len(old_result.get('issues', [])))
    new_issues = new_result.get('issues', []) + [''] * (max_issues - len(new_result.get('issues', [])))
    
    issues_df = pd.DataFrame({
        "Legacy Script Issues": old_issues,
        "Enhanced Script Issues": new_issues
    })
    
    display(issues_df)
    
    # 5. USER AGENT ANALYSIS TABLE
    if 'ua_analysis' in new_result:
        print("\nüë§ USER AGENT DISCRIMINATION ANALYSIS")
        print("-" * 50)
        
        ua_data = []
        for ua_name, data in new_result['ua_analysis'].items():
            status = data.get('status', 0)
            status_text = f"{status} {'üî¥' if status in [403, 429, 503] else 'üü°' if status != 200 else 'üü¢'}"
            
            ua_data.append({
                "User Agent": ua_name.replace('_', ' ').title(),
                "Status": status_text,
                "Response Time": f"{data.get('elapsed', 0):.2f}s",
                "Content Length": data.get('headers', {}).get('Content-Length', 'N/A')
            })
        
        ua_df = pd.DataFrame(ua_data)
        display(ua_df)
        
        # UA discrimination summary
        ua_analysis = new_result.get('detailed_analysis', {}).get('ua_discrimination', {})
        print(f"\nüìä Discrimination Summary:")
        print(f"   Bot Blocked: {'üî¥ YES' if ua_analysis.get('bot_blocked') else 'üü¢ NO'}")
        print(f"   Headless Blocked: {'üî¥ YES' if ua_analysis.get('headless_blocked') else 'üü¢ NO'}")
        print(f"   Response Time Variance: {ua_analysis.get('response_time_variance', 0):.3f}")
    
    # 6. BOT PROTECTION SERVICES DETECTED
    bot_protection = new_result.get('detailed_analysis', {}).get('bot_protection', {})
    if bot_protection:
        print("\nüõ°Ô∏è BOT PROTECTION SERVICES DETECTED")
        print("-" * 50)
        
        for service_id, service_data in bot_protection.items():
            confidence = service_data.get('confidence', 'unknown').upper()
            risk = service_data.get('risk_level', 'unknown').upper()
            print(f"   üî¥ {service_data['service_name']}: {confidence} confidence, {risk} risk")
            print(f"      Evidence: {len(service_data.get('content_matches', []))} content + {len(service_data.get('header_matches', []))} header matches")
    else:
        print("\nüõ°Ô∏è BOT PROTECTION: üü¢ No advanced services detected")
    
    # 7. RECOMMENDATIONS COMPARISON
    print("\nüí° RECOMMENDATIONS COMPARISON")
    print("-" * 50)
    
    max_recs = max(len(old_result.get('recommendations', [])), len(new_result.get('recommendations', [])), 1)
    old_recs = old_result.get('recommendations', []) + [''] * (max_recs - len(old_result.get('recommendations', [])))
    new_recs = new_result.get('recommendations', []) + [''] * (max_recs - len(new_result.get('recommendations', [])))
    
    recs_df = pd.DataFrame({
        "Legacy Recommendations": old_recs,
        "Enhanced Recommendations": new_recs
    })
    
    display(recs_df)
    
    # 8. IMPROVEMENT SUMMARY
    print("\nüéâ IMPROVEMENT SUMMARY")
    print("-" * 50)
    
    improvements = [
        "‚úÖ Multi-User-Agent Differential Testing (6 profiles)",
        "‚úÖ Advanced Bot Protection Detection (8 services)", 
        "‚úÖ Comprehensive Fingerprinting Analysis (12+ patterns)",
        "‚úÖ Rate Limiting Detection with Severity Scoring",
        "‚úÖ JavaScript Challenge Detection",
        "‚úÖ Intelligent robots.txt Parsing", 
        "‚úÖ Granular 0-100 Scoring System",
        "‚úÖ Parallel Execution for Performance",
        "‚úÖ Production-Ready Error Handling",
        "‚úÖ Detailed Technical Analysis"
    ]
    
    for improvement in improvements:
        print(f"   {improvement}")
    
    print("\n" + "="*90)
    print("‚ú® COMPARISON COMPLETE - ENHANCED SCRIPT READY FOR PRODUCTION")
    print("="*90)


# ============================================================================
# DEMONSTRATION AND TESTING
# ============================================================================

def run_demonstration():
    """
    üéØ DEMONSTRATION FUNCTION - Test the enhanced module
    
    Runs both legacy and enhanced versions on a test URL and displays
    comprehensive comparison results.
    """
    print("üöÄ ANTI-AUTOMATION READINESS MODULE DEMONSTRATION")
    print("="*60)
    
    # Get URL from user
    url = input("üåê Enter URL to analyze (or press Enter for default): ").strip()
    if not url:
        url = "https://www.google.com"
        print(f"   Using default URL: {url}")
    
    print(f"\nüîç Analyzing: {url}")
    print("="*60)
    
    try:
        # Run Legacy Version
        print("\nüî¥ RUNNING LEGACY SCRIPT...")
        html = requests.get(url, timeout=10).text
        old_score, old_issues, old_recommendations = check_anti_automation_absence(url, html)
        
        old_result = {
            "url": url,
            "score": old_score,
            "readiness": "LIMITED" if old_score >= 3 else "BLOCKED",
            "issues": old_issues,
            "recommendations": old_recommendations
        }
        
        print(f"   Legacy Score: {old_score}/5")
        
        # Run Enhanced Version  
        print("\nüü¢ RUNNING ENHANCED SCRIPT...")
        new_result = check_anti_automation_readiness_enhanced(url)
        print(f"   Enhanced Score: {new_result['score']}/100 - {new_result['classification']}")
        
        # Display comparison
        visualize_anti_automation_comparison(old_result, new_result)
        
        return old_result, new_result
        
    except Exception as e:
        print(f"‚ùå Demonstration failed: {str(e)}")
        return None, None


# ============================================================================
# MAIN EXECUTION BLOCK
# ============================================================================

if __name__ == "__main__":
    """
    üéØ MAIN EXECUTION - Run when script is executed directly
    
    This block runs the demonstration when the script is executed directly,
    but doesn't run when the module is imported.
    """
    print("üöÄ ENHANCED ANTI-AUTOMATION READINESS MODULE")
    print("Version 2.0 - Production Ready")
    print("="*50)
    
    # Run demonstration
    old_result, new_result = run_demonstration()
    
    # Optional: Save results to file
    if new_result and input("\nüíæ Save results to JSON file? (y/n): ").lower() == 'y':
        filename = f"anti_automation_analysis_{time.strftime('%Y%m%d_%H%M%S')}.json"
        with open(filename, 'w') as f:
            json.dump(new_result, f, indent=2)
        print(f"‚úÖ Results saved to: {filename}")

# ============================================================================
# USAGE EXAMPLES AND INTEGRATION GUIDE
# ============================================================================

"""
üéØ QUICK START USAGE EXAMPLES:

# 1. Basic usage - Enhanced version
result = check_anti_automation_readiness_enhanced("https://example.com")
print(f"Score: {result['score']}/100")
print(f"Classification: {result['classification']}")

# 2. Legacy compatibility (for existing code)
score, issues, recommendations = check_anti_automation_absence("https://example.com")

# 3. Batch analysis
urls = ["https://site1.com", "https://site2.com"]
for url in urls:
    analysis = check_anti_automation_readiness_enhanced(url)
    if analysis['score'] >= 60:
        print(f"‚úÖ {url}: Ready for automation")
    else:
        print(f"‚ùå {url}: Automation barriers detected")

# 4. Integration with ARI framework
def automation_audit(url):
    anti_auto = check_anti_automation_readiness_enhanced(url)
    return {
        "automation_readiness": anti_auto,
        "score": anti_auto["score"],
        "classification": anti_auto["classification"],
        "recommendations": anti_auto["recommendations"]
    }
"""

"""
üìä OUTPUT STRUCTURE (Enhanced Version):

{
    "url": "https://example.com",
    "score": 85,  # 0-100 scale
    "classification": "READY",  # READY/LIMITED/RESTRICTED/BLOCKED
    "readiness_level": "Agent-friendly",
    "issues": ["List of identified issues"],
    "recommendations": ["Actionable recommendations"],
    "execution_time": 2.34,
    "timestamp": "2024-01-15 10:30:45",
    "detailed_analysis": {
        "ua_discrimination": { ... },
        "bot_protection": { ... },
        "fingerprinting": { ... },
        "rate_limiting": { ... },
        "robots_txt": { ... },
        "javascript_analysis": { ... }
    },
    "ua_analysis": { ... },  # For visualization
    "js_checks": { ... }     # For visualization
}
"""

"""
üîß MIGRATION GUIDE FROM LEGACY:

1. REPLACE:
   check_anti_automation_absence(url, html) 
   WITH:
   check_anti_automation_readiness_enhanced(url)

2. UPDATE SCORE PROCESSING:
   Old: 0-5 scale ‚Üí New: 0-100 scale

3. ENHANCE REPORTING:
   Use the comprehensive issues and recommendations lists

4. LEVERAGE DETAILED ANALYSIS:
   Access technical details in detailed_analysis for advanced use cases
"""
````
