```python
# fibbanic series 
def fibbanic_series(a,b, n):
    lst = [a, b]
    for i in range(2, n):
        c = a + b
        a = b
        b = c
        lst.append(c)


    return lst



# function call
n = int(input('Enter values of n : '))
a = int(input('Enter values of a : '))
b = int(input('Enter values of b : '))

print(fibbanic_series(a, b, n))

        
    
```

```python
def fibbanic_series(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    
    # Start with first two numbers
    series = [0, 1]
    
    # Loop from 2 to n
    for i in range(2, n):
        next_val = series[i-1] + series[i-2]
        series.append(next_val)
        
    return series

# function call
n = int(input('Enter number of terms : '))
print(fibbanic_series(n))

    
```
