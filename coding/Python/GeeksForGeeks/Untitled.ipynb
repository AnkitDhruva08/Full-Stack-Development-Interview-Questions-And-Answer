```python
class Student:
    school = "ABC School"

    def __init__(self, name):
        self.name = name

    def show_name(self):        # instance method
        print(self.name)

    @classmethod
    def school_name(cls):      # class method
        print(cls.school)

    @staticmethod
    def info():                # static method
        print("Students belong to a school")



# main program 
s = Student('Ankit')
print(s.show_name())
print(s.school_name())
print(s.info())
```

```python
import asyncio
import time

async def request(name, delay):
    print(f"{name} started")
    await asyncio.sleep(delay)
    print(f"{name} finished")

async def main():
    start = time.time()

    task1 = asyncio.create_task(request("Request-1", 5))
    task2 = asyncio.create_task(request("Request-2", 2))

    await asyncio.gather(task1, task2)

    print("Total time:", time.time() - start)

await main()
```

```python
import time
import asyncio

async def blocking_request(name, delay):
    print(f"{name} started")
    time.sleep(delay)   # blocking
    print(f"{name} finished")

async def main():
    start = time.time()

    task1 = asyncio.create_task(blocking_request("Request-1", 5))
    task2 = asyncio.create_task(blocking_request("Request-2", 2))

    await asyncio.gather(task1, task2)

    print("Total time:", time.time() - start)

await main()
```

```python
import asyncio
import time

async def task(name):
    for i in range(3):
        print(f"{name} running {i}")
        await asyncio.sleep(1)

async def main():
    start = time.time()

    t1 = asyncio.create_task(task("A"))
    t2 = asyncio.create_task(task("B"))

    await asyncio.gather(t1, t2)

    print("Total time:", time.time() - start)

await main()
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass


class Dog(Animal):

    def sound(self):
        return "Bark"


class Cat(Animal):

    def sound(self):
        return "Meow"


dog = Dog()
print(dog.sound())
```

```python
def decorator_fun(func):
    def wrapper():
        print('Hello Ankit')
        func()
        print("After function call")

    return wrapper


@decorator_fun
def say_hello():
    print("Hello")
    return "Done"



say_hello()
```
