```python
# Sum of Digits of a Number
def Sum_of_Digits(n):
    digit_sum = 0

    while n != 0:
        remain = n % 10
        digit_sum = digit_sum + remain
        n //= 10
   
    return digit_sum



# function call
n = int(input("enter a number : "))
print(Sum_of_Digits(n))
```

```python
# Primality Test and School Method Algorithms
def isPrime(n):
    isPrime = False
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            isPrime = False
            break
    else:
        isPrime = True

    return isPrime
    
            

# function call
n = int(input("enter a number : "))
print(isPrime(n))
        
```

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)


# function call
n = int(input("enter a number : "))
print(factorial(n))
```

```python
def gcd(a, b):
  
    # Everything divides 0
    if a == 0:
        return b
    if b == 0:
        return a

    # Base case
    if a == b:
        return a

    # a is greater
    if a > b:
        return gcd(a - b, b)
    return gcd(a, b - a)

if __name__ == '__main__':
    a = 20
    b = 28
    print(gcd(a, b))
```

```python
def fib_series(n):
    if n == 1:
        return [0]

    if n == 2:
        return [0,1]

    for i in range(n):
        lst = fib_series(n - 1)
        lst.append(lst[-1] + lst [-2])

    return lst


print(fib_series(5))
```
