```python
TIME COMPLEXITY :
    Time complexity measures how the execution time of an algorithm increases as the input size increases.
    Time Complexity is the rate of growth of an algorithm’s running time with respect to the input size.

️ Why do we need Time Complexity?
    1. To compare algorithms
    
    2. To predict performance for large inputs
    
    3. To choose the most efficient solution
    
    4. Hardware-independent analysis
What is Input Size (n)?
    n = number of elements / operations / data size
    1. Array length → n

    2. Number of nodes in tree → n

    3. Characters in string → n

What does Time Complexity measure?
    1. Number of basic operations

    2. As input size n increases

Big-O Notation (Most Important)
    Big-O represents the upper bound (worst-case) time complexity.

    Common Big-O values:
        | Complexity | Meaning      |
        | ---------- | ------------ |
        | O(1)       | Constant     |
        | O(log n)   | Logarithmic  |
        | O(n)       | Linear       |
        | O(n log n) | Linearithmic |
        | O(n²)      | Quadratic    |
        | O(2ⁿ)      | Exponential  |

Best, Average, Worst Case
    | Case    | Meaning                           |
    | ------- | --------------------------------- |
    | Best    | Minimum time                      |
    | Average | Expected time                     |
    | Worst   | Maximum time (Big-O focuses here) |


Time Complexity vs Space Complexity
    | Time                    | Space                 |
    | ----------------------- | --------------------- |
    | Measures execution time | Measures memory usage |
    | Focus on speed          | Focus on memory       |
```

```python
Big-O Notation
    Big-O notation describes the upper bound (worst-case) growth rate of an algorithm’s time or space complexity
    as the input size increases.

    Big-O notation expresses an asymptotic upper bound on the growth rate of an algorithm’s running time or space complexity
    as a function of input size.
        
How bad can it get when input becomes very large?

ideal case should be : omega
average case should be : theeta

What Big-O actually does
Big-O:

    1. Measures growth rate, not actual time

    2. Focuses on worst-case scenario

    3. Ignores:

        1. constants

        2. lower-order terms

        3. Used for comparison, not measurement


Big-O vs Big-Θ vs Big-Ω (Quick correction)
    | Notation | Meaning                  |
    | -------- | ------------------------ |
    | Big-O    | Worst-case (upper bound) |
    | Big-Θ    | Exact bound              |
    | Big-Ω    | Best-case (lower bound)  |
```

```python
# Loop with Constant Work
for i in range(n):
    print(i)

# O(n)
```

```python
# Nested Loops → Multiply
# Nested loops
for i in range(n):
    for j in range(n):
        print(i, j)
# Total → n × n = n²
# O(n²)
```

```python
# Different ranges
for i in range(n):
    for j in range(m):
        print(i, j)

# O(n × m)
```

```python
# Consecutive Loops → Add, Then Simplify
for i in range(n):
    print(i)

for j in range(n):
    print(j)
# Time: n + n = 2n
# Ignore constants
# O(n)
```

```python
# Drop Constants & Lower Terms (MOST IMPORTANT RULE)
for i in range(n):
    for j in range(n):
        print(i)

for k in range(n):
    print(k)
# Time:n² + n
# Keep highest growth
# O(n²)
```
