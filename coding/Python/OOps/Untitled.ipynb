```python
from abc import ABC, abstractmethod


class BankAccount(ABC):
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance   # 🔒 Encapsulation (private)

    # 🔹 Encapsulation: controlled access
    def get_balance(self):
        return self.__balance

    def _update_balance(self, amount):
        self.__balance += amount

    # 🔹 Abstraction: force implementation
    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass


class SavingsAccount(BankAccount):

    def deposit(self, amount):
        if amount <= 0:
            return "Invalid deposit amount"
        
        self._update_balance(amount)
        return f"{amount} deposited. Balance: {self.get_balance()}"

    def withdraw(self, amount):
        if amount <= 0:
            return "Invalid withdrawal amount"
        
        if amount > self.get_balance():
            return "Insufficient balance"
        
        self._update_balance(-amount)
        return f"{amount} withdrawn. Balance: {self.get_balance()}"


class CreditAccount(BankAccount):
    def __init__(self, name, balance, credit_limit):
        super().__init__(name, balance)
        self.credit_limit = credit_limit

    def deposit(self, amount):
        if amount <= 0:
            return "Invalid deposit amount"
        
        self._update_balance(amount)
        return f"Payment received: {amount}. Balance: {self.get_balance()}"

    def withdraw(self, amount):
        if amount <= 0:
            return "Invalid amount"
        
        if amount > (self.get_balance() + self.credit_limit):
            return "Credit limit exceeded"
        
        self._update_balance(-amount)
        return f"Spent {amount}. Balance: {self.get_balance()}"
```

```python
# Savings account
acc1 = SavingsAccount("Ankit", 1000)
print(acc1.deposit(500))
print(acc1.withdraw(200))

# Credit account
acc2 = CreditAccount("Rahul", 0, 5000)
print(acc2.withdraw(3000))   # allowed (credit)
print(acc2.withdraw(3000))   # exceeds limit
```

```python
import threading
import time

def task(name):
    print(f"Start {name}")
    time.sleep(2)
    print(f"End {name}")

t1 = threading.Thread(target=task, args=("T1",))
t2 = threading.Thread(target=task, args=("T2",))

t1.start()
t2.start()

t1.join()
t2.join()

print("Done")
```

```python
#genex1.py
def  kvrrange(k,v,r):
	while(k<=v):
		yield k
		k=k+r

#main program
r=range(10,20,2)
print("type of r=",type(r))
print(r)
for v in r:
	print(v)
print("--------------------------------------")
kvrobj=kvrrange(100,120,5)
print("type of kvrobj=",type(kvrobj))
while(True):
	try:
		print(next(kvrobj))
	except StopIteration:
		break
```
