```python
def palindrom_index(s):
    if s == s[::-1]:
        return -1

    result = []
    indices = []

    for i in range(len(s)):
        for j in range(i + 1, len(s) + 1):
            sub = s[i:j]
            if len(sub) > 1 and sub == sub[::-1]:
                result.append(sub)
                indices.append(i)
    for i in indices:
        return i


s = input("")
print(palindrom_index(s))

        
```

```python
def palindrom_index(s):
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] == s[right]:
            left += 1
            right -= 1 

        else:
            # try romving left charecter
            if s[left+1:right+1] == s[left+1:right+1][::-1]:
                return left 

            # Try removing right character
            if s[left:right] == s[left:right][::-1]:
                return right 

    return -1
                
            
```

```python
# two string 
def twoStrings(s1, s2):
    for i in s1:
        for j in s2:
            if ord(i) == ord(j):
                return "YES"

    return "NO"


n = int(input(''))
for i in range(n):
    s1 = input('')
    s2 = input('')
    print(twoStrings(s1, s2))
            
```

```python
def twoStrings(s1, s2):
    if set(s1) & set(s2):
        return "YES"
    return "NO"


n = int(input())
for _ in range(n):
    s1 = input().strip()
    s2 = input().strip()
    print(twoStrings(s1, s2))
```

```python
# def isValid(s):
def isValid(s):
    freq = {}

    # Count characters
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    # Count frequencies of frequencies
    freq_count = {}
    for v in freq.values():
        freq_count[v] = freq_count.get(v, 0) + 1

    # Case 1: all same frequency
    if len(freq_count) == 1:
        return "YES"

    # Case 2: two different frequencies
    if len(freq_count) == 2:
        keys = list(freq_count.keys())
        f1, f2 = keys[0], keys[1]

        # One char with frequency 1
        if (f1 == 1 and freq_count[f1] == 1) or (f2 == 1 and freq_count[f2] == 1):
            return "YES"

        # One char with frequency one more than others
        if abs(f1 - f2) == 1:
            if freq_count[max(f1, f2)] == 1:
                return "YES"

    return "NO"


s = input().strip()
print(isValid(s))
```
