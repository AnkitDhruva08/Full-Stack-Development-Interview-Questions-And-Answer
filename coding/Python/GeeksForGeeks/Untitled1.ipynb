```python
class Solution:
    def maxProfit(self, arr, k):
        if not arr:
            return 0
        cash = 0
        hold = -arr[0]
        print('hold :', hold)
        
        for i in range(1, len(arr)):
            prev_cash = cash
            print('prev_cash :', prev_cash)
            
            cash = max(cash, hold + arr[i] - k)
            print('cash :', cash)
            
            hold = max(hold, prev_cash - arr[i])
            print('hold :', hold)
            
            
        return cash

# function call 
s = Solution()
arr = list(map(int,input("entyer a list values : ").split(',')))
k = int(input("enter k values : "))
print(s.maxProfit(arr, k))
```
