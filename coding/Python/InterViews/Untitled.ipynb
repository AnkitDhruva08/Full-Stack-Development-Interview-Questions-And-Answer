```python
class Animal:
    def speak(self):
        return "Animal can speak"

class Cat(Animal):
    def speak(self):
        return "Meow"


a = Animal()
c = Cat()
print(a.speak())
print(c.speak())
```

```python
def overload(a, b, *args):
    return a + b + sum(args)

print(overload(3, 2))        # 5
print(overload(3, 2, 5))     # 10
print(overload(3, 2, 5, 7))  # 17
```

```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)


print(gcd(12, 18))
    
```
