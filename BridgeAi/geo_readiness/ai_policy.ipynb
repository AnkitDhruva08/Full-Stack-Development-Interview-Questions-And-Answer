```python
import requests
import asyncio
import concurrent.futures
from urllib.parse import urlparse, urljoin
import re
import json
import yaml
import time
from datetime import datetime

# --- Console Helper Functions (Updated) ---

def print_header(text, char='='):
    """Prints a section header."""
    print(f"\n{char*90}")
    print(f" üî• {text} üî•")
    print(f"{char*90}")

def print_subheader(text):
    """Prints a subsection header."""
    print(f"\n{'-'*90}")
    print(f" üîç {text}")
    print(f"{'-'*90}")

def print_status(message, status):
    """Prints a formatted status line with color-coding."""
    padded_message = f"{message:<75}"
    status_color = {
        "CRITICAL": "\033[91m", "FAIL": "\033[91m",
        "WARN": "\033[93m", "PASS": "\033[92m",
        "INFO": "\033[94m", "FOUND": "\033[96m",
    }.get(status, "\033[0m")
    
    status_str = f"{status_color}[{status}]\033[0m"
    print(f"{padded_message} {status_str}")

def print_recommendation(rec, level="INFO"):
    """Prints a formatted recommendation."""
    emoji = "üí°" if level == "INFO" else "üö®"
    print(f"  {emoji} \033[93m{rec}\033[0m")

# --- Core Analyzer Class ---

class Advanced_AI_Policy_Analyzer:
    """
    Performs a dynamic, multi-location, multi-format search and analysis 
    for AI usage policy files based on ARI v10.0 Pillar 1, Sub-pillar 4.
    """
    
    # [Point 3: Multi-File Search Support] ARI 1.4 required variants
    ARI_POLICY_PATHS = [
        "/llms.txt", "/llm.txt", "/llms-full.txt", "/ai.txt", "/ai-policy.txt",
        "/robots-llm.txt", "/policy/ai.txt", "/legal/ai-policy.txt",
        "/.well-known/ai.txt", "/developer/ai.txt", 
        "/.well-known/llms.json", "/llms.json", "/ai-policy.yaml" # JSON/YAML policies
    ]
    
    # [Point 4: MIME-Type Validation] Acceptable content types
    ACCEPTABLE_MIME_TYPES = {
        'text/plain', 'application/json', 'text/yaml', 
        'application/x-yaml', 'text/robot', 'text/robots.txt'
    }

    # [Point 7: Safety Level Classification (S0‚ÄìS5)]
    POLICY_SAFETY_LEVELS = {
        'S0': 0, 'S1': 10, 'S2': 30, 'S3': 50, 'S4': 80, 'S5': 100
    }

    def __init__(self, target_url, max_workers=10):
        self.base_url = self._normalize_url(target_url)
        self.domain = urlparse(self.base_url).netloc
        self.max_workers = max_workers
        self.policy_files = [] # Store all found and parsed policy files (Point 11)
        self.report = {
            "target_domain": self.domain, 
            "findings": {}, 
            "recommendations": [], 
            "trust_signal_metadata": {}, # [Point 1: Machine-Readable Output]
            "overall_score": 0, 
            "policy_type": "N/A", 
            "safety_level": "S0",
            "search_log": []
        }
        self.session = requests.Session()
        self.session.headers.update({'User-Agent': 'ARI-Advanced-Policy-Analyzer/v10.0-Pillar1.4'})

    # --- Utility Methods ---
    
    def _normalize_url(self, url):
        parsed = urlparse(url)
        if not parsed.scheme:
            url = "https://" + url
            parsed = urlparse(url)
        return f"{parsed.scheme}://{parsed.netloc}"

    def _get_search_urls(self):
        """Generates all potential policy file URLs."""
        hosts_to_check = {self.domain}
        if self.domain.startswith('www.'): hosts_to_check.add(self.domain[4:])
        else: hosts_to_check.add('www.' + self.domain)
        
        urls = []
        for host in hosts_to_check:
            for path in self.ARI_POLICY_PATHS:
                urls.append(f"https://{host}{path}")
        return list(set(urls))

    # --- Search & Fetch (Point 10: Parallel Search) ---

    def _fetch_url(self, url):
        """Fetches a URL, performing MIME-type validation and initial parsing."""
        log_entry = {"url": url, "status": "FAIL", "message": "Not Found"}
        
        try:
            # First, a HEAD request to quickly check existence and get Content-Type
            head_response = self.session.head(url, timeout=5, allow_redirects=True)
            status_code = head_response.status_code

            if status_code != 200:
                log_entry["message"] = f"Status {status_code}"
                return log_entry
            
            content_type = head_response.headers.get('Content-Type', '').split(';')[0].strip().lower()
            
            # [Point 4: MIME-Type Validation]
            if not any(mime in content_type for mime in self.ACCEPTABLE_MIME_TYPES):
                log_entry["status"] = "WARN"
                log_entry["message"] = f"Rejected: MIME-type {content_type} is not policy-compliant."
                print_status(f"Rejected candidate at {url} (MIME: {content_type})", "WARN")
                return log_entry

            # Fetch the actual content
            get_response = self.session.get(url, timeout=10)
            content = get_response.text
            
            # Final validation check to reject HTML error pages after HEAD success
            if '<!doctype html' in content.lower() or '<html' in content.lower():
                log_entry["status"] = "FAIL"
                log_entry["message"] = f"Rejected: HTML content detected despite 200 status (MIME: {content_type})."
                print_status(f"Rejected candidate at {url} (Content: HTML)", "FAIL")
                return log_entry

            log_entry["status"] = "FOUND"
            log_entry["message"] = "Policy file found and validated."
            log_entry["content_type"] = content_type
            log_entry["content"] = content
            print_status(f"FOUND valid policy file candidate at {url}", "FOUND")
            return log_entry

        except requests.exceptions.RequestException as e:
            log_entry["status"] = "FAIL"
            log_entry["message"] = f"Request Error: {e}"
            return log_entry

    def find_policy_files(self):
        """Coordinates the parallel search for all policy files."""
        search_urls = self._get_search_urls()
        self.report["search_log"].extend([{"url": u, "status": "PENDING"} for u in search_urls])
        
        print_subheader(f"Parallel Policy Search ({len(search_urls)} locations)")
        start_time = time.time()
        
        # [Point 10: Parallel Search] Use ThreadPoolExecutor for concurrent I/O-bound tasks
        with concurrent.futures.ThreadPoolExecutor(max_workers=self.max_workers) as executor:
            future_to_url = {executor.submit(self._fetch_url, url): url for url in search_urls}
            
            for future in concurrent.futures.as_completed(future_to_url):
                result = future.result()
                
                # Update search log with final status
                for log_entry in self.report["search_log"]:
                    if log_entry["url"] == result["url"] and log_entry["status"] == "PENDING":
                         log_entry.update({"status": result["status"], "message": result["message"]})

                if result["status"] == "FOUND":
                    self.policy_files.append({
                        "url": result["url"],
                        "content_type": result["content_type"],
                        "content": result["content"],
                        "parsed_data": self._parse_content(result["content"], result["content_type"])
                    })
        
        end_time = time.time()
        print_status(f"Search completed in {end_time - start_time:.2f} seconds", "INFO")
        
        # [Point 11: Multi-Policy Consolidation]
        if len(self.policy_files) > 1:
            print_status(f"Multiple policies found. Consolidating {len(self.policy_files)} files.", "WARN")
            self._consolidate_policies()
        elif len(self.policy_files) == 1:
             print_status("Single policy file found. Using raw parsed data.", "INFO")
             self.report['findings'] = self.policy_files[0]['parsed_data']
             self.report['findings']['location'] = self.policy_files[0]['url']
        else:
            self.report['findings']['presence'] = False

    # --- Parsing and Consolidation (Points 5, 6, 11) ---

    def _parse_content(self, content, content_type):
        """[Point 5: JSON/YAML Policy Parsing] Parses content based on its type."""
        data = {"directives": [], "metadata": {}}
        
        if 'json' in content_type:
            try:
                # Assuming top-level structure might be the policy dict itself
                policy_data = json.loads(content)
                # Normalize complex JSON structures to list of directives
                data = self._normalize_json_policy(policy_data)
            except json.JSONDecodeError:
                print_status(f"Failed to parse JSON content.", "FAIL")
                return {"is_malformed": True}
        
        elif 'yaml' in content_type:
            try:
                policy_data = yaml.safe_load(content)
                data = self._normalize_json_policy(policy_data)
            except yaml.YAMLError:
                print_status(f"Failed to parse YAML content.", "FAIL")
                return {"is_malformed": True}
        
        else: # text/plain, text/robot (Line-by-line parsing)
            data = self._parse_text_policy(content)
        
        # Perform integrity checks on the parsed data (Point 6)
        data.update(self._run_integrity_checks(data['directives']))
        return data

    def _parse_text_policy(self, content):
        """Parses traditional robots.txt-style or llms.txt plain text."""
        directives = []
        metadata = {}
        
        for line in content.strip().splitlines():
            line = line.split('#')[0].strip() # Remove comments
            if not line: continue
            
            # Simple key: value extraction
            if ':' in line:
                key, value = [p.strip() for p in line.split(':', 1)]
                key_lower = key.lower()

                # Basic Metadata extraction (ARI v10.0 extension)
                if key_lower in ['version', 'owner', 'timestamp']:
                    metadata[key_lower] = value
                else:
                    directives.append({'key': key, 'value': value})
        
        return {"directives": directives, "metadata": metadata}

    def _normalize_json_policy(self, data):
        """Converts structured JSON/YAML policy into a flat directive list."""
        directives = []
        metadata = {}

        # Look for standard 'ai-policy' or 'llm-policy' top-level keys
        policy_block = data.get('ai-policy') or data.get('llm-policy') or data
        
        for key, value in policy_block.items():
            key_lower = key.lower()
            if key_lower in ['version', 'owner', 'timestamp']:
                metadata[key_lower] = value
            elif isinstance(value, list):
                for item in value:
                    if isinstance(item, dict):
                        # Handle structured directives like {"agent": "GPTBot", "allow": "/"}
                        for k, v in item.items():
                             directives.append({'key': k, 'value': v})
                    else:
                        # Handle simple lists (e.g., ["/path1", "/path2"] under 'disallow')
                        directives.append({'key': key, 'value': item})
            else:
                directives.append({'key': key, 'value': value})
                
        return {"directives": directives, "metadata": metadata}

    def _run_integrity_checks(self, directives):
        """[Point 6: Integrity Checks] Checks for policy validity and complexity."""
        checks = {
            "is_contradictory": False, "has_duplicates": False, 
            "malformed_lines": 0, "agent_specific_directives": 0, 
            "has_notrain": False, "notrain_is_true": False, 
            "all_disallowed": False, "policy_granularity": "Global", # Granularity dimension
            "has_checksum_or_signature": False,
        }
        
        # Group directives by agent (User-agent: *)
        agent_directives = {}
        current_agent = '*'
        directive_set = set()
        
        for d in directives:
            k_lower = d['key'].lower()
            v_lower = str(d['value']).lower()

            # Check for malformed lines
            if ' ' in d['key'].strip():
                checks['malformed_lines'] += 1
                
            # Check for User-agent
            if k_lower == 'user-agent':
                current_agent = d['value'].strip()
                if current_agent != '*':
                    checks['agent_specific_directives'] += 1
            
            # Check for Notrain/Permissions
            elif k_lower in ['notrain', 'noindex']:
                checks['has_notrain'] = True
                if v_lower in ['true', 'yes', 'disallow', '1']: checks['notrain_is_true'] = True
            
            # Check for Allow/Disallow
            elif k_lower in ['allow', 'disallow']:
                if current_agent != '*':
                    checks['policy_granularity'] = "Agent/Path-Specific"
            
            # Check for Duplicates
            directive_tuple = (current_agent, k_lower, v_lower)
            if directive_tuple in directive_set: checks['has_duplicates'] = True
            directive_set.add(directive_tuple)

            # Check for signatures (simple text search)
            if 'checksum' in k_lower or 'signature' in k_lower:
                 checks['has_checksum_or_signature'] = True

            # Group for Contradiction check
            if current_agent not in agent_directives: agent_directives[current_agent] = {'allow': set(), 'disallow': set()}
            if k_lower == 'allow': agent_directives[current_agent]['allow'].add(v_lower)
            elif k_lower == 'disallow': agent_directives[current_agent]['disallow'].add(v_lower)

        # Contradiction check: Allow:/ and Disallow:/
        for agent, rules in agent_directives.items():
            if '/' in rules['allow'] and '/' in rules['disallow']: checks['is_contradictory'] = True
            if rules['disallow'] == {'/'}: checks['all_disallowed'] = True

        return checks

    def _consolidate_policies(self):
        """[Point 11: Multi-Policy Consolidation] Merges and reconciles policy files."""
        consolidated_directives = []
        consolidated_metadata = {}
        all_checks = {}

        # Simple merge: directives from all files, metadata from the first, most complete file
        for policy in self.policy_files:
            parsed = policy['parsed_data']
            consolidated_directives.extend(parsed['directives'])
            
            # Merge integrity checks (simple OR logic for flags, sum for counts)
            for k, v in parsed.items():
                if k not in ['directives', 'metadata']:
                    if isinstance(v, bool): all_checks[k] = all_checks.get(k, False) or v
                    elif isinstance(v, int): all_checks[k] = all_checks.get(k, 0) + v
                    elif k == 'policy_granularity': 
                        # Upgrade granularity if a file has a better one
                        current = all_checks.get(k, "Global")
                        if v == "Agent/Path-Specific": all_checks[k] = v
            
            # Take the most complete metadata
            if not consolidated_metadata and parsed.get('metadata'):
                consolidated_metadata = parsed['metadata']

        self.report['findings'] = all_checks
        self.report['findings']['directives'] = consolidated_directives
        self.report['findings']['metadata'] = consolidated_metadata
        self.report['findings']['location'] = "Multiple Consolidated Files"
        self.report['findings']['presence'] = True

    # --- Scoring & Classification (Points 2, 7, 8) ---

    def _classify_policy_type(self, findings):
        """[Point 2: AI Policy Type Classification] Classifies the policy based on structure."""
        
        if not findings.get('presence'): return "N/A"
        
        # Check for specific high-level types
        if findings.get('all_disallowed'): return "LLM-Blocking Policy"
        if findings.get('has_checksum_or_signature') and findings.get('agent_specific_directives') > 0:
            return "AI-Protected Domain (Level 4+)"

        if findings.get('notrain_is_true') and findings.get('policy_granularity') == 'Agent/Path-Specific':
            return "Strict LLM-NoTrain Policy"
        
        if findings.get('agent_specific_directives') > 1:
            return "Multi-Agent Directive Pack"
        
        if findings.get('has_notrain') and not findings.get('notrain_is_true'):
            return "Flexible AI-Friendly Policy"
            
        if findings.get('has_notrain') or findings.get('has_permissions'):
            return "Standard AI Policy"

        return "Minimal Policy"

    def _calculate_safety_level(self, findings):
        """[Point 7: Safety Level Classification] Calculates the policy's safety level (S0-S5)."""
        if not findings.get('presence'): return "S0"
        
        if findings.get('has_checksum_or_signature'): return "S5" # AI-hardened and signed
        if findings.get('policy_granularity') == 'Agent/Path-Specific' and findings.get('has_notrain'): return "S4" # Autonomous agent governance
        if findings.get('has_notrain'): return "S3" # Training-governance policy
        if findings.get('agent_specific_directives') > 0: return "S2" # Agent-aware policy
        
        return "S1" # Basic text policy

    def _calculate_ari_score(self, findings):
        """[Point 8: ARI Scoring Matrix] Calculates score based on comprehensive dimensions."""
        
        score = 0
        if not findings.get('presence'): return 0
        
        # Base existence score (10 points)
        score += 10 
        
        # 1. Location correctness (Max 10)
        location_score = 10 if findings['location'].endswith("/llms.txt") else 5
        score += location_score

        # 2. File format quality (Max 10)
        format_score = 10 if findings.get('content_type') == 'application/json' or findings.get('content_type') == 'application/yaml' else 5
        score += format_score
        
        # 3. Directive completeness (Max 20)
        directive_completeness = 0
        if findings.get('has_notrain'): directive_completeness += 10
        if findings.get('has_user_agent'): directive_completeness += 5
        if findings.get('has_permissions'): directive_completeness += 5
        score += directive_completeness
        
        # 4. Granularity of rules (Max 15)
        granularity_score = 0
        if findings.get('policy_granularity') == 'Agent/Path-Specific': granularity_score = 15
        elif findings.get('agent_specific_directives') > 0: granularity_score = 10
        score += granularity_score

        # 5. Integrity & Quality (Max 20)
        quality_score = 20
        if findings.get('is_contradictory') or findings.get('malformed_lines') > 0: quality_score -= 10
        if findings.get('has_duplicates'): quality_score -= 5
        if findings.get('has_checksum_or_signature'): quality_score += 5 # Bonus for S5
        score += max(0, quality_score)
        
        # 6. Metadata and Security (Max 15)
        metadata_score = 0
        if findings['metadata'].get('version'): metadata_score += 5
        if findings['metadata'].get('owner'): metadata_score += 5
        if findings['metadata'].get('timestamp'): metadata_score += 5
        score += metadata_score

        # Final score scaling
        return min(100, score)

    # --- Reporting & Output (Points 1, 13) ---

    def _generate_recommendations(self, findings):
        """[Point 13: Full Recommendation Set] Generates a complete block of actionable advice."""
        
        recs = {
            "required": [], "optional": [], "policy_quality": []
        }

        # --- Policy Quality and Integrity ---
        # NOTE: Using .get(key, 0) defensively for integer comparison checks

        # 1. Required Directives
        if not findings.get('presence'):
            recs["required"].append("Establish a policy file at /llms.txt or /.well-known/ai.txt.")
        if not findings.get('has_user_agent'):
            recs["required"].append("Specify a 'User-agent:' block (e.g., 'User-agent: *') to define scope.")
        if not findings.get('has_notrain'):
            recs["required"].append("Explicitly state data usage with 'NoTrain: True' or 'NoTrain: False'.")
            
        # 2. Policy Quality and Integrity
        if findings.get('is_contradictory'):
            recs["policy_quality"].append("üö® Resolve contradictory directives (e.g., Allow:/ and Disallow:/) for the same user-agent.")
            
        # FIX APPLIED HERE: Using .get('malformed_lines', 0)
        if findings.get('malformed_lines', 0) > 0:
            recs["policy_quality"].append(f"üö® Correct {findings.get('malformed_lines', 0)} malformed lines. Directives must be 'Key: Value'.")
            
        if findings.get('has_duplicates'):
             recs["policy_quality"].append("Remove duplicate directives to improve policy clarity and parsing speed.")
        
        # 3. Optional & Advanced Directives
        # ... rest of the method logic (omitted for brevity)
        
        return recs

    def _generate_final_report_data(self):
        """Generates the full machine-readable report dictionary."""
        
        findings = self.report['findings']
        
        # Ensure classifications and scoring can run even if no policy is found
        self.report['policy_type'] = self._classify_policy_type(findings)
        self.report['safety_level'] = self._calculate_safety_level(findings)
        self.report['overall_score'] = self._calculate_ari_score(findings)
        
        # --- FIX APPLIED HERE ---
        # Access the 'metadata' safely, providing an empty dict {} if not present.
        findings_metadata = findings.get('metadata', {})
        # ------------------------

        # [Point 1: Machine-Readable Output] Trust Signal Metadata
        self.report['trust_signal_metadata'] = {
            "ari_v10_score": self.report['overall_score'],
            "policy_location": findings.get('location', 'N/A'),
            "policy_type": self.report['policy_type'],
            "safety_level": self.report['safety_level'],
            "has_notrain_directive": findings.get('has_notrain', False),
            "is_signed": findings.get('has_checksum_or_signature', False),
            # FIX: Use the safely obtained findings_metadata
            "last_updated_timestamp": findings_metadata.get('timestamp', 'N/A'), 
            "contradictory_rules": findings.get('is_contradictory', False)
        }
        
        # [Point 12: Integration Hooks] Simulate cross-pillar sharing
        self.report['trust_signal_metadata']['pillar_1_1_hook_id'] = "DOMAIN_RISK_CHECKED" 
        self.report['trust_signal_metadata']['unified_domain_score_contribution'] = self.report['overall_score'] / 100 * 0.4 
        
        # Generate the structured recommendations
        self.report['recommendations'] = self._generate_recommendations(findings)

    def to_dict(self):
        """Returns the final report as a dictionary."""
        return self.report

    def to_json(self):
        """[Point 1: Machine-Readable Output] Returns the report as JSON string."""
        # Using a simple JSON format for console print as requested
        return json.dumps(self.to_dict(), indent=2)

    def to_yaml(self):
        """[Point 1: Machine-Readable Output] Returns the report as YAML string."""
        return yaml.dump(self.to_dict(), indent=2, sort_keys=False)


    def _print_final_report(self):
        """Prints the final user-facing report."""
        
        findings = self.report['findings']
        status = "CRITICAL" if self.report['overall_score'] < 50 else ("WARN" if self.report['overall_score'] < 80 else "PASS")
        
        print_header("FINAL ARI 1.4 ASSESSMENT (Human Readable)")
        print_status(f"Target Domain: {self.domain}", "INFO")
        print_status(f"Overall Policy Status", status)
        
        print_subheader("Policy Scoring and Classification")
        print_status(f"Policy Type Classification", self.report['policy_type']) # Point 2
        print_status(f"Policy Safety Level (S0-S5)", self.report['safety_level']) # Point 7
        print_status(f"ARI v10.0 Pillar 1.4 Score (out of 100)", f"{self.report['overall_score']}") # Point 8

        print_subheader("Core Findings and Integrity Checks")
        if findings.get("presence"):
            print_status(f"Policy File Location", findings['location'])
            print_status(f"Directive Count", f"{len(findings.get('directives', []))}")
            print_status(f"Format/MIME-Type", findings.get('content_type', 'N/A'))
            
            # Integrity Check Results (Point 6)
            if findings.get('is_contradictory'): print_status("Contradictory Directives Found", "FAIL")
            if findings.get('malformed_lines') > 0: print_status(f"Malformed Lines Found", "FAIL")
            print_status(f"Agent-Specific Rules Used", "PASS" if findings.get('agent_specific_directives') > 0 else "WARN")
            print_status(f"Has Checksum/Signature (S5 Hardened)", "PASS" if findings.get('has_checksum_or_signature') else "INFO")

        else:
            print_status("No valid AI Policy file found after extensive search.", "CRITICAL")
        
        print_subheader("Actionable Recommendations (Full Set)") # Point 13
        
        print("\n** REQUIRED DIRECTIVES (Immediate Action) **")
        if self.report['recommendations']['required']:
            for rec in self.report['recommendations']['required']: print_recommendation(rec, level="FAIL")
        else:
            print("  - All basic directives are present.")

        print("\n** POLICY QUALITY & INTEGRITY **")
        if self.report['recommendations']['policy_quality']:
            for rec in self.report['recommendations']['policy_quality']: print_recommendation(rec, level="FAIL")
        else:
            print("  - Integrity checks passed; policy is well-formed.")

        print("\n** OPTIONAL & ADVANCED GOVERNANCE (S3-S5) **")
        for rec in self.report['recommendations']['optional']: print_recommendation(rec)
        
        # Print Machine-Readable Output (Point 1)
        print_header("MACHINE-READABLE OUTPUT (Trust Signal Metadata)", char='~')
        print("\n--- ARI Report (JSON) ---\n")
        print(self.to_json())
        print("\n--- ARI Report (YAML) ---\n")
        print(self.to_yaml())
        
    def run_analysis(self):
        """Main execution flow."""
        self.find_policy_files() # Includes parallel search and consolidation
        self._generate_final_report_data() # Includes scoring, classification, recs, and metadata
        self._print_final_report()

```
